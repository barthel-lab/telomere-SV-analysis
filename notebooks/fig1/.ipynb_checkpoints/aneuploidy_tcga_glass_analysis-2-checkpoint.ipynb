```r
# ---- Load genomic arm coordinates ----
arm_lengths <- read_tsv("genomic_arm_coordinates_hg19.txt", col_types = cols()) %>%
  rename(chrom = chrom, arm = arm, arm_start = start, arm_end = end) %>%
  mutate(arm_length = arm_end - arm_start)
head(arm_lengths)
```

```r


genome_size <- sum(arm_lengths$arm_length)

# Load CNV data as before
files <- list.files(pattern = "_(primary|recurrence).seg$")
cnv_all <- do.call(rbind, lapply(files, function(file) {
  df <- read.table(file, header = TRUE, sep = "\t")
  df$tumor_type <- case_when(
    grepl("astro", file, ignore.case = TRUE) ~ "Astrocytoma",
    grepl("oligo", file, ignore.case = TRUE) ~ "Oligodendroglioma",
    grepl("glioblastoma", file, ignore.case = TRUE) ~ "GBM",
    TRUE ~ "Other"
  )
  df$stage <- ifelse(grepl("primary", file, ignore.case = TRUE), "Primary", "Recurrence")
  df$sample <- df$ID
  df$row_id <- paste0(file, "_", seq_len(nrow(df)))  # unique id
  df
}))

cnv_all <- cnv_all %>%
  mutate(chrom = ifelse(grepl("^chr", chrom), chrom, paste0("chr", chrom)),
         segment_length = loc.end - loc.start)

# Join with arm coordinates & calculate overlap
cnv_with_arms <- cnv_all %>%
  left_join(arm_lengths, by = "chrom") %>%
  filter(loc.start < arm_end & loc.end > arm_start) %>%
  mutate(
    overlap_start = pmax(loc.start, arm_start),
    overlap_end = pmin(loc.end, arm_end),
    overlap_len = pmax(0, overlap_end - overlap_start),
    arm_fraction = overlap_len / arm_length
  )
head(cnv_with_arms)
```

```r
library(tidyverse)
library(ComplexHeatmap)
library(circlize)
library(cowplot)

# --- Assuming these data frames exist from your previous steps: ---
# arm_level_cnvs (columns: sample, tumor_type, stage, region, cna_status)
# partial_cnvs (same columns)
# measures (sample, tumor_type, stage, aneuploidy_fraction, percent_genome_altered)

# --- Step 1: Prepare matrices for heatmaps ---

arm_level_mat <- arm_level_cnvs %>%
  group_by(sample, region) %>%
  summarise(cna_status = first(cna_status), .groups = "drop") %>%
  pivot_wider(names_from = region, values_from = cna_status, values_fill = 0) %>%
  column_to_rownames("sample") %>%
  as.matrix()

partial_mat <- partial_cnvs %>%
  group_by(sample, region) %>%
  summarise(cna_status = first(cna_status), .groups = "drop") %>%
  pivot_wider(names_from = region, values_from = cna_status, values_fill = 0) %>%
  column_to_rownames("sample") %>%
  as.matrix()

# --- Step 2: Get common samples to match rows in both matrices ---

common_samples <- intersect(rownames(arm_level_mat), rownames(partial_mat))

arm_level_mat_sub <- arm_level_mat[common_samples, , drop = FALSE]
partial_mat_sub <- partial_mat[common_samples, , drop = FALSE]

# --- Helper function to get tumor type in correct order ---
get_tumor_type <- function(samples, df) {
  df %>%
    distinct(sample, tumor_type) %>%
    filter(sample %in% samples) %>%
    arrange(factor(sample, levels = samples)) %>%
    pull(tumor_type)
}

tumor_colors <- c("GBM" = "#E41A1C", "Astrocytoma" = "#377EB8", "Oligodendroglioma" = "#4DAF4A")

# --- Step 3: Create row annotations for subset samples ---

ha_arm_row_sub <- rowAnnotation(
  TumorType = factor(get_tumor_type(common_samples, arm_level_cnvs),
                     levels = names(tumor_colors)),
  col = list(TumorType = tumor_colors),
  show_annotation_name = TRUE
)

ha_partial_row_sub <- rowAnnotation(
  TumorType = factor(get_tumor_type(common_samples, partial_cnvs),
                     levels = names(tumor_colors)),
  col = list(TumorType = tumor_colors),
  show_annotation_name = TRUE
)

# --- Step 4: Define CNA color function ---
col_fun <- colorRamp2(c(-1, 0, 1), c("darkblue", "lightyellow1", "orangered"))
```

```r
# Step 1: Define desired chromosome arm order (1p, 1q, ..., Xp, Xq)
arm_order <- paste0(rep(c(1:22, "X", "Y"), each = 2), rep(c("p", "q"), times = 24))

# Step 2: Clean up column names to 1p, 1q, etc.
clean_colnames <- function(cols) {
  cols <- gsub("^chr(\\d+)_\\1", "\\1", cols)  # chr7_7p → 7p
  cols <- gsub("^chrX_X", "X", cols)
  cols <- gsub("^chrY_Y", "Y", cols)
  return(cols)
}

# Apply to both matrices
colnames(arm_level_mat_sub) <- clean_colnames(colnames(arm_level_mat_sub))
colnames(partial_mat_sub) <- clean_colnames(colnames(partial_mat_sub))

# Step 3: Reorder columns based on desired arm order
# Keep only columns that exist in the matrix and in our defined order
valid_arm_cols <- intersect(arm_order, colnames(arm_level_mat_sub))
arm_level_mat_sub <- arm_level_mat_sub[, valid_arm_cols]

valid_partial_cols <- intersect(arm_order, colnames(partial_mat_sub))
partial_mat_sub <- partial_mat_sub[, valid_partial_cols]
```

```r
# Get tumor type factor in the order of the rows
row_split_factor <- factor(get_tumor_type(common_samples, arm_level_cnvs),
                           levels = c("GBM", "Astrocytoma", "Oligodendroglioma"))

options(repr.plot.width = 15, repr.plot.height = 10, repr.plot.res = 200)

ht_arm_sub <- Heatmap(arm_level_mat_sub,
                      name = "Arm-level CNA",
                      col = col_fun,
                      right_annotation = ha_arm_row_sub,
                      show_row_names = FALSE,
                      cluster_rows = FALSE,
                      cluster_columns = FALSE,
                      column_title = "Arm-level CNAs",
                      row_split = row_split_factor,
                      row_title = NULL,   # optionally hide row titles if you want
                      cluster_row_slices = TRUE)  # cluster rows within each slice

ht_partial_sub <- Heatmap(partial_mat_sub,
                         name = "Partial CNA",
                         col = col_fun,
                         right_annotation = ha_partial_row_sub,
                         show_row_names = FALSE,
                         cluster_rows = FALSE,
                         cluster_columns = FALSE,
                         column_title = "Partial CNAs",
                         row_split = row_split_factor,
                         row_title = NULL,
                         cluster_row_slices = TRUE)
```

```r
#––– Libraries
library(dplyr)
library(ggplot2)
library(ggpubr)
library(forcats)

#––– Parameters (edit these two)
arm_threshold     <- 0.8      # fixed cutoff for Arm‑level (≥ this → Arm-level)
partial_threshold >- 0.25      # variable cutoff for Partial‑level (< this → Partial-level)

amp_cutoff  <- 0.25            # |seg.mean| cutoff for any CNV
genome_size <- 3.117e9        # T2T human genome length (bp)

#––– Prep
df <- cnv_with_arms %>%
  mutate(
    seg.mean     = as.numeric(seg.mean),
    arm_fraction = as.numeric(arm_fraction)
  )

#––– Classify
df2 <- df %>%
  filter(abs(seg.mean) > amp_cutoff) %>%
  mutate(
    category = case_when(
      arm_fraction >= arm_threshold      ~ "Arm-level",
      arm_fraction <  partial_threshold  ~ "Partial-level",
      TRUE                                ~ NA_character_
    )
  ) %>%
  filter(!is.na(category))

#––– Summarize per sample & category
cnv_summary <- df2 %>%
  group_by(sample, tumor_type, category) %>%
  summarise(
    altered_bp = sum(overlap_len, na.rm=TRUE),
    .groups = "drop"
  ) %>%
  mutate(
    proportion = altered_bp / genome_size,
    tumor_type = factor(tumor_type,
                        levels = c("Astrocytoma","GBM","Oligodendroglioma"))
  )

#––– Colors & comparisons
my_cols <- c(
  "GBM"              = "#E41A1C",
  "Astrocytoma"      = "#377EB8",
  "Oligodendroglioma"= "#4DAF4A"
)
my_comparisons <- list(
  c("Astrocytoma","GBM"),
  c("Astrocytoma","Oligodendroglioma"),
  c("GBM","Oligodendroglioma")
)

#––– Plot function
plot_cnv <- function(cat_name, y_lab){
  dat <- cnv_summary %>%
    filter(category == cat_name) %>%
    mutate(
      # reorder x by ascending median
      tumor_type = fct_reorder(tumor_type, proportion, median)
    )
  
  ggplot(dat, aes(tumor_type, proportion,
                  fill = tumor_type, color = tumor_type)) +
    geom_violin(trim = FALSE, scale = "width", alpha = 0.6) +
    geom_boxplot(width = 0.2, fill = "white", outlier.shape = NA) +
    scale_fill_manual(values = my_cols) +
    scale_color_manual(values = my_cols) +
    stat_compare_means(
      method      = "wilcox.test",
      comparisons = my_comparisons,
      label       = "p.signif"
    ) +
    labs(x = "", y = y_lab) +
    theme_classic() +
    theme(
      legend.position = "none",
      axis.text       = element_text(size = 12),
      axis.title      = element_text(size = 13)
    )
}
```

```r
#––– Draw
p1 <- plot_cnv("Arm-level",     "Proportion altered (Arm‑level CNVs)")
p2 <- plot_cnv("Partial-level", "Proportion altered (Partial‑level CNVs)")
```

```r
#––– Libraries
library(dplyr)
library(ggplot2)
library(ggforce)    # for geom_sina()
library(ggpubr)     # for stat_compare_means()
library(scales)     # for percent_format()

#––– Parameters
amplitude_threshold <- 0.01                    # |seg.mean| cutoff
genome_length       <- 3054815472              # T2T‑CHM13 nuclear genome size (bp)

#––– 1) Filter and compute per‑sample CNV burden
cnv_burden <- cnv_with_arms %>%
  filter(
    cna_status != 0,                           # only gains/losses
    abs(seg.mean) >= amplitude_threshold       # amplitude threshold
  ) %>%
  group_by(sample, tumor_type, stage) %>%
  summarise(
    raw_bp      = sum(overlap_len),            # total bases altered
    weighted_bp = sum(abs(seg.mean) * overlap_len),  # amplitude‑weighted
    .groups     = "drop"
  ) %>%
  mutate(
    pct_raw      = raw_bp      / genome_length,   # raw % of genome
    pct_weighted = weighted_bp / genome_length    # weighted % of genome
  )

#––– 2) Define Primary vs Recurrence comparison
stage_cmp <- list(c("Primary", "Recurrence"))

#––– 3) Plot amplitude‑weighted % genome altered
ggplot(cnv_burden, aes(x = stage, y = pct_weighted, fill = stage)) +
  
  # violin + white boxplot + sina points
  geom_violin(trim = FALSE, alpha = 0.4) +
  geom_boxplot(width = 0.3, fill = "white", outlier.shape = NA) +
  geom_sina(aes(color = stage), size = 1.2, alpha = 0.6, maxwidth = 0.6) +
  
  # add Wilcoxon test p‑values per facet
  stat_compare_means(
    comparisons = stage_cmp,
    method      = "wilcox.test",
    label       = "p.signif",
    tip.length  = 0.02,
    label.y     = max(cnv_burden$pct_weighted) * 0.5
  ) +
  
  # y axis as percent
  scale_y_continuous(
    labels = percent_format(accuracy = 0.1),
    limits = c(0, NA),
    expand = c(0, 0.02)
  ) +
  
  # facet by tumor_type, free y‑scale so each subtype shows its own range
  facet_wrap(~ tumor_type, nrow = 1, scales = "free_y") +
  
  # labels and theme
  labs(
    x     = NULL,
    y     = "Amplitude‑weighted % of Genome Altered",
    title = paste0(
      "CNV Burden (|seg.mean| ≥ ", amplitude_threshold,
      ") by Stage and Tumor Type"
    )
  ) +
  theme_bw() +
  theme(
    legend.position = "none",
    strip.text      = element_text(size = 13, face = "bold"),
    axis.text       = element_text(size = 11),
    axis.title      = element_text(size = 13),
    plot.title      = element_text(size = 15, face = "bold", hjust = 0.5)
  )
```

```r
#––– Libraries
library(dplyr)
library(ggpubr)
library(scales)

#––– PARAMETERS
arm_threshold     <- 0.8      # ≥ this → Arm‑level
partial_threshold <- 0.3      # < this → Partial‑level
amp_cutoff        <- 0.3      # |seg.mean| cutoff
genome_size       <- 3.117e9  # T2T genome size (bp)
tumor_type_filter   <- "Oligodendroglioma"      # <— change as needed

#––– PREP + CLASSIFY
df2 <- cnv_with_arms %>%
  mutate(
    seg.mean     = as.numeric(seg.mean),
    arm_fraction = as.numeric(arm_fraction)
  ) %>%
  filter(
    abs(seg.mean) > amp_cutoff,
    tumor_type == tumor_type_filter
  ) %>%
  mutate(category = case_when(
    arm_fraction >= arm_threshold     ~ "Arm‑level",
    arm_fraction >  partial_threshold ~ "Partial‑level",
    TRUE                               ~ NA_character_
  )) %>%
  filter(!is.na(category)) %>%
  group_by(sample, stage, category) %>%
  summarise(
    altered_bp = sum(overlap_len, na.rm=TRUE),
    .groups    = "drop"
  ) %>%
  mutate(
    proportion = altered_bp / genome_size,
    stage      = factor(stage, levels = c("Primary","Recurrence"))
  )

#––– Plot size (notebook)
options(repr.plot.width = 8, repr.plot.height = 6, repr.plot.res = 300)

#––– PAIRED PLOTS WITH ggpaired + STATISTICS
ggpaired(df2,
         x          = "stage",
         y          = "proportion",
         id         = "sample",
         color      = "stage",
         line.color = "grey50",
         line.type  = "dashed",
         palette    = c("#377EB8","#E41A1C"),
         facet.by   = "category",
         slope.linetype = "dashed",
         legend     = "none"
) +
  # add paired Wilcoxon test p‑values per facet
  stat_compare_means(
    method      = "wilcox.test",
    paired      = FALSE,
    comparisons = list(c("Primary","Recurrence")),
    label       = "p.signif",
    tip.length  = 0.02
  ) +
  scale_y_continuous(labels = percent_format(accuracy = 1)) +
  labs(
    x     = "",
    y     = "Proportion altered (genome‑wide)",
    title = paste0(tumor_type_filter, ": Primary vs Recurrence")
  ) +
  theme_classic() +
  theme(
    strip.text = element_text(face="bold", size=12),
    axis.text  = element_text(size=11),
    axis.title = element_text(size=13),
    plot.title = element_text(size=14, hjust=0.5)
  )
```

```r
#––– Libraries
library(dplyr)
library(tidyr)
library(ggpubr)
library(scales)

#––– PARAMETERS
arm_threshold     >- 0.8      # ≥ this → Arm‑level
partial_threshold < 0.8      # < this → Partial‑level
amp_cutoff        >- 0.01      # |seg.mean| cutoff
#tumor_type_filter <- "Astrocytoma"    # change as needed
#tumor_type_filter   <- "Oligodendroglioma"
tumor_type_filter   <- "Astrocytoma"
df_counts_total <- cnv_with_arms %>%
  mutate(
    seg.mean     = as.numeric(seg.mean),
    arm_fraction = as.numeric(arm_fraction)
  ) %>%
  filter(
    abs(seg.mean) > amp_cutoff,
    tumor_type == tumor_type_filter
  ) %>%
  mutate(category = case_when(
    arm_fraction >= arm_threshold     ~ "Arm‑level",
    arm_fraction <  partial_threshold ~ "Partial‑level",
    TRUE                               ~ NA_character_
  )) %>%
  filter(!is.na(category)) %>%
  # count across all arms
  group_by(sample, stage, category) %>%
  summarise(cnv_count = n(), .groups="drop") %>%
  mutate(
    stage    = factor(stage, levels = c("Primary","Recurrence")),
    category = factor(category, levels = c("Arm‑level","Partial‑level"))
  )

#––– Plot size (if in a notebook)
options(repr.plot.width = 8, repr.plot.height = 6, repr.plot.res = 300)

#––– PAIRED PLOT: total CNVs per sample with log10 y‐axis
ggpaired(df_counts_total,
         x          = "stage",
         y          = "cnv_count",
         id         = "sample",
         color      = "stage",
         palette    = c("#377EB8","#E41A1C"),
         line.color = "grey50",
         line.type  = "dashed",
         line.size  = 0.8,
         facet.by   = "category",
         legend     = "none"
) +
  # paired Wilcoxon test p‐values per facet
  stat_compare_means(
    method      = "wilcox.test",
    paired      = FALSE,
    comparisons = list(c("Primary","Recurrence")),
    label       = "p.signif",
    tip.length  = 0.02
  ) +
  # log10 scale on y; add nice breaks & log‐ticks
  scale_y_log10(
    breaks = trans_breaks("log10", function(x) 10^x),
    labels = trans_format("log10", math_format(10^.x))
  ) +
  annotation_logticks(sides = "l") +
  labs(
    x     = NULL,
    y     = "log10(Total CNV segments)",
    title = paste0(tumor_type_filter, ": Total CNVs Primary vs Recurrence")
  ) +
  theme_bw() +
  theme(
    strip.text = element_text(face="bold", size=12),
    axis.text  = element_text(size=11),
    axis.title = element_text(size=13),
    plot.title = element_text(size=14, hjust=0.5)
  )
```

```r
#––– Libraries
library(dplyr)
library(ggplot2)
library(ggpubr)

#––– PARAMETERS
arm_threshold >- 0.8       # ≥ this → Arm‑level event
amp_cutoff    <- 0.01      # |seg.mean| cutoff
tumor_types   <- c("Astrocytoma", "GBM", "Oligodendroglioma")



options(repr.plot.width = 10, repr.plot.height = 4, repr.plot.res = 300)

#––– Function to compute & plot for one tumor type
plot_aneu <- function(tumor) {
  df <- cnv_with_arms %>%
    mutate(
      seg.mean     = as.numeric(seg.mean),
      arm_fraction = as.numeric(arm_fraction),
      stage        = factor(stage, levels = c("Primary", "Recurrence"))
    ) %>%
    filter(
      tumor_type == tumor,
      abs(seg.mean) > amp_cutoff,
      arm_fraction >= arm_threshold
    ) %>%
    distinct(sample, stage, arm) %>%      # one count per altered arm
    group_by(sample, stage) %>%
    summarise(aneuploidy_score = n(), .groups="drop")
  
  ggpaired(
    df,
    x         = "stage",
    y         = "aneuploidy_score",
    id        = "sample",
    color     = "stage",
    line.color= "gray",
    line.size = 0.4,
    palette   = c("#377EB8","#E41A1C"),
    point.size= 1.5,
    ylab      = "Aneuploidy score\n(# altered arms)",
    xlab      = ""
  ) +
    stat_compare_means(
      paired = FALSE,
      method = "t.test",
      label  = "p.signif"
    ) +
    theme_bw() +
    ggtitle(paste0(tumor, ": Primary vs. Recurrence"))
}

#––– Generate one plot per tumor type
plots <- lapply(tumor_types, plot_aneu)

#––– Display them all in a single column
ggarrange(
  plotlist = plots,
  ncol     = 3,
  nrow     = 1,
  align    = "v"
)
```
