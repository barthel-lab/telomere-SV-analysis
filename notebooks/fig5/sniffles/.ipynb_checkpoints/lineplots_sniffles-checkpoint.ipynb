```r
library(dplyr)
library(tidyr)
library(ggplot2)
library(ComplexHeatmap)
library(circlize)
```

```r
# Read the CSV file into a data frame
#df_cnv <- read.delim("Mean_normalized_v3_tab.seg", header = TRUE)
df_arms <- read.delim("cytoBandMapped_v2.bed", header = TRUE)
t2t_telomere = read.csv(file = "chm13v2.0_telomere.bed_v2.txt", header = TRUE, sep = "\t")
# Set column names
```

```r
#head(df_cnv)
head(df_arms)
head(t2t_telomere)
```

```r
#assign p and q arms to t2t telomeres data
t2t_telomere <- t2t_telomere %>%
  #rename(chrom = V1, telomere_start = V2, telomere_end = V3) %>%
  group_by(chr) %>%
  arrange(start) %>%
  mutate(Arm = case_when(
    row_number() == 1 ~ "p",
    row_number() == n() ~ "q",
    TRUE ~ NA_character_
  )) %>%
  ungroup()
head(t2t_telomere)
```

```r
colnames(df_arms) <- c("chr", "start", "end", "band", "stain")
head(df_arms)
```

```r
# Step 2: Create 'arm' column
df_arms <- df_arms %>%
  mutate(arm = ifelse(
    stain == "acen", "centromere",
    ifelse(grepl("^p", band), "p", "q")
  ))

# Step 3: Calculate the size for each row
df_arms <- df_arms %>%
  mutate(size = end - start)

# Step 4: Group by chromosome and arm, and calculate total size, start, and end
df_arms <- df_arms %>%
  group_by(chr, arm) %>%
  summarise(
    total_size = sum(size),
    start = min(start),  # get the first start position for each group
    end = max(end)       # get the last end position for each group
  )

# Step 5: View the results
print(df_arms)
```

```r
# Merge the data
t2t_merged_data <- bind_rows(
  df_arms %>% select(chr, arm, start, end),
  t2t_telomere %>% mutate(arm = paste0(Arm, "_telomere")) %>% select(chr, arm, start, end)
) %>%
  arrange(chr, start)

# View the merged data
head(t2t_merged_data)
```

```r
sv_data = read.csv(file = "SVs_merged_all_samples_SVs_v7.txt", sep = '\t', header = TRUE)
head(sv_data)
```

```r
# First, fix types for start/end in sv_data_clean
sv_data_clean <- sv_data %>%
  mutate(
    start = as.numeric(start),
    end = as.numeric(end)
  )
```

```r
head(sv_data_clean)
head(t2t_merged_data)
```

```r
unique(sv_data_clean$type)
```

```r
# Join separate arm coordinates safely
sv_scaled <- sv_data_clean %>%
  select(-arm) %>%  # remove existing 'arm' column if present
  left_join(
    df_arms %>% filter(arm == "p") %>% select(chr, start_p = start), 
    by = "chr"
  ) %>%
  left_join(
    df_arms %>% filter(arm == "q") %>% select(chr, end_q = end), 
    by = "chr"
  ) %>%
  left_join(
    df_arms %>% filter(arm == "centromere") %>% select(chr, centromere_start = start), 
    by = "chr"
  ) %>%
  mutate(
    midpoint = (start + end) / 2,
    scaled_pos = (midpoint - start_p) / (end_q - start_p),
    centromere_pos = if_else(
      !is.na(centromere_start) & !is.na(start_p) & !is.na(end_q),
      (centromere_start - start_p) / (end_q - start_p),
      NA_real_
    )
  ) %>%
  filter(!is.na(scaled_pos))
```

```r
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 1) load libraries
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
library(tidyverse)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 2) filter to chr9
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
sv9  <- sv_data_clean    %>% filter(chr == "chr9")
seg9 <- t2t_merged_data  %>% filter(chr == "chr9")

# total length of chr9
chr9_len <- max(seg9$end)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 3) define telomeric regions for shading
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
telomeres9 <- seg9 %>% filter(arm %in% c("p_telomere", "q_telomere"))

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 4) build 1 Mb bins along chr9
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
bin_size <- 1e6
bins9 <- tibble(
  bin   = 1:ceiling(chr9_len / bin_size),
  start = (bin - 1) * bin_size,
  end   = pmin(bin * bin_size, chr9_len)
)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 5) assign each SV to a bin by its start coordinate
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
sv9_binned <- sv9 %>% 
  mutate(bin = floor(start / bin_size) + 1) %>% 
  filter(bin <= nrow(bins9))

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 6) count SVs per type × bin; fill missing with 0
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
counts9 <- sv9_binned %>% 
  count(type, bin) %>% 
  complete(
    type = unique(sv9_binned$type),
    bin  = 1:nrow(bins9),
    fill = list(n = 0)
  ) %>% 
  rename(count = n) %>% 
  left_join(bins9, by = "bin") %>% 
  mutate(
    density    = count / (end - start),
    enrichment = density / (nrow(sv9) / chr9_len),
    midpos_mb  = (start + end) / (2 * 1e6)
  )

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 7) exclude BND calls
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
counts9_filt <- counts9 %>% filter(type != "BND")

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 8) define dual‐axis plotting function
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
plot_dual <- function(df, types, cols, title) {
  sub <- df %>% filter(type %in% types)
  
  # scale factor so enrichment × sf ≈ count
  sf <- max(sub$count, na.rm = TRUE) / max(sub$enrichment, na.rm = TRUE)
  
  ggplot() +
    # telomere shading
    geom_rect(
      data        = telomeres9,
      inherit.aes = FALSE,
      aes(
        xmin = start/1e6,
        xmax = end  /1e6,
        ymin = -Inf,
        ymax = +Inf
      ),
      fill  = "lightyellow",
      alpha = 0.4
    ) +
    # enrichment (solid, prominent)
    geom_line(
      data = sub,
      aes(x = midpos_mb, y = enrichment, color = type),
      size  = 1.5,
      alpha = 1
    ) +
    # count (solid, subtler, rescaled)
    geom_line(
      data = sub,
      aes(x = midpos_mb, y = count / sf, color = type),
      size  = 0.6,
      alpha = 0.5
    ) +
    # left axis = enrichment, right axis = count
    scale_y_continuous(
      name    = "Enrichment (obs/exp density)",
      sec.axis = sec_axis(
        ~ . * sf,
        name = "SV count per Mb"
      )
    ) +
    scale_color_manual(values = cols) +
    labs(
      x     = "chr9 position (Mb)",
      title = title,
      color = "SV type"
    ) +
    theme_bw() +
    theme(
      legend.position   = "bottom",
      panel.grid.minor  = element_blank()
    )
}

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 9) generate the two plots with custom colors
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
p_ins_del <- plot_dual(
  counts9_filt,
  types = c("INS", "DEL"),
  cols  = c(INS = "palegreen2", DEL = "steelblue2"),
  title = "INS & DEL: Enrichment (solid) vs Count (solid)"
)

p_inv_dup <- plot_dual(
  counts9_filt,
  types = c("INV", "DUP"),
  cols  = c(INV = "orange3",   DUP = "tomato"),
  title = "INV & DUP: Enrichment (solid) vs Count (solid)"
)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 10) display
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
print(p_ins_del)
print(p_inv_dup)
```

```r
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 1) load libraries
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
library(tidyverse)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 2) parameters & SV types
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
bin_size <- 1e6
sv_types <- c("INS","DEL","INV","DUP")

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 3) chromosome lengths + cumulative offsets, drop chrX & chrY
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
chr_lengths <- t2t_merged_data %>%
  group_by(chr) %>%
  summarise(chr_len = as.numeric(max(end)), .groups = "drop") %>%
  mutate(chr_no = parse_number(chr)) %>%
  arrange(!is.na(chr_no), chr_no, chr) %>%
  mutate(cum_start = lag(cumsum(chr_len), default = 0)) %>%
  filter(!chr %in% c("chrX","chrY"))

# for x‐axis labels
chr_midpoints <- chr_lengths %>%
  transmute(chr,
            mid_mb = (cum_start + chr_len/2) / 1e6)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 4) build genome‐wide 1 Mb bins (autosomes only)
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
bins_all <- chr_lengths %>%
  mutate(
    n_bins = ceiling(chr_len / bin_size),
    bin    = map(n_bins, seq_len)
  ) %>%
  unnest(bin) %>%
  mutate(
    bin_start        = (bin - 1) * bin_size,
    bin_end          = pmin(bin * bin_size, chr_len),
    midpos           = (bin_start + bin_end) / 2,
    global_midpos_mb = (cum_start + midpos) / 1e6
  ) %>%
  select(chr, bin, bin_start, bin_end, global_midpos_mb)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 5) telomere & centromere shading (autosomes only)
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
shade_all <- t2t_merged_data %>%
  filter(arm %in% c("p_telomere","centromere","q_telomere")) %>%
  filter(chr %in% chr_lengths$chr) %>%     # drop X/Y
  left_join(chr_lengths, by = "chr") %>%
  mutate(
    xmin_mb = (cum_start + start) / 1e6,
    xmax_mb = (cum_start + end)   / 1e6
  ) %>%
  select(chr, arm, xmin_mb, xmax_mb)

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 6) assign SVs to those bins & count
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
sv_binned <- sv_data_clean %>%
  filter(type %in% sv_types, chr %in% chr_lengths$chr) %>%  # drop BND & X/Y
  mutate(bin = floor(start / bin_size) + 1) %>%
  inner_join(bins_all, by = c("chr","bin"))

genome_len   <- sum(chr_lengths$chr_len)
total_svs    <- nrow(sv_binned)
expected_den <- total_svs / genome_len

expanded <- expand_grid(type = sv_types, bins_all)

counts_all <- expanded %>%
  left_join(
    sv_binned %>%
      count(type, chr, bin, global_midpos_mb, bin_start, bin_end, name="count"),
    by = c("type","chr","bin","global_midpos_mb","bin_start","bin_end")
  ) %>%
  replace_na(list(count = 0)) %>%
  mutate(
    density    = count / (bin_end - bin_start),
    enrichment = density / expected_den
  )

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 7) dual‐axis genome‐wide plotting function (with larger base font)
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
plot_dual_genome <- function(df, types, cols, title) {
  sub <- df %>% filter(type %in% types)
  sf  <- max(sub$count) / max(sub$enrichment)
  
  ggplot() +
    # yellow shading for telomeres & centromeres
    geom_rect(
      data        = shade_all,
      inherit.aes = FALSE,
      aes(xmin = xmin_mb, xmax = xmax_mb, ymin = -Inf, ymax = Inf),
      fill  = "yellow2",
      alpha = 0.7
    ) +
    # enrichment = thick solid
    geom_line(
      data = sub,
      aes(x = global_midpos_mb, y = enrichment, color = type),
      size = 1.5, alpha = 1
    ) +
    # raw count = thin solid
    geom_line(
      data = sub,
      aes(x = global_midpos_mb, y = count / sf, color = type),
      size  = 1.5, alpha = 0.6
    ) +
    scale_y_continuous(
      name     = "Enrichment (obs/exp SVs)",
      sec.axis = sec_axis(~ . * sf, name = "SV count per Mb")
    ) +
    scale_x_continuous(
      breaks = chr_midpoints$mid_mb,
      labels = chr_midpoints$chr,
      expand = c(0, 0)
    ) +
    scale_color_manual(values = cols) +
    labs(
      x     = "Chromosomes",
      title = title,
      color = "SV type"
    ) +
    theme_bw(base_size = 34) +                        # ← bigger base font
    theme(
      axis.text.x      = element_text(angle = 90, vjust = 0.5),
      legend.position  = "bottom",
      panel.grid.minor = element_blank()
    )
}

#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
# 8) build & show the two genome‐wide plots
#––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––––
p_ins_del_genome <- plot_dual_genome(
  counts_all,
  types = c("INS","DEL"),
  cols  = c(INS = "palegreen2", DEL = "steelblue2"),
  title = "INS & DEL across the Genome"
)

p_inv_dup_genome <- plot_dual_genome(
  counts_all,
  types = c("INV","DUP"),
  cols  = c(INV = "orange2", DUP = "tomato"),
  title = "INV & DUP across the Genome"
)
```
