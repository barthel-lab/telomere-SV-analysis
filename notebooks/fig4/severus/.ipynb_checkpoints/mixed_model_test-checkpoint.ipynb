```r
# Mixed-effects model overview (why we use it here)
# We have repeated measurements per patient/sample (many chr_arms per Sample_ID), so observations are not independent.
# A mixed model accounts for this by modeling Sample_ID as a random effect (patient-specific baseline), while estimating the fixed TL effect (and allowing TL slopes to differ by chr_arm via TL * chr_arm).
# We then extract the per-arm TL slope and its p-value for each genomic feature to build matrices for heatmap visualization (either raw p-values for exploration or BH-FDR for multiple-testing control).
```

```r
library(dplyr)
library(lme4)
library(lmerTest)  # IMPORTANT: load this AFTER lme4
library(emmeans)
```

```r
#This function is testing for the mixed model, calculating FDR

# What mixed_corr_mats() does
# 1) For each genomic feature column, it builds a clean analysis table with TL (chosen column), the feature (Y), chr_arm, and Sample_ID, dropping missing values and optionally filtering arms with small n.
# 2) It fits a mixed model: Y ~ TL * chr_arm + (1 | Sample_ID), which estimates arm-specific TL slopes while accounting for non-independence of arms within the same Sample_ID.
# 3) It uses emtrends() to extract the TL slope per chr_arm, converts the slope t-statistic to an r-like signed effect size, and returns two matrices: r (effect size) and p (raw p-values) aligned as feature × chr_arm for heatmap plotting.



mixed_corr_mats <- function(df, tl_col, features, arms, min_n_per_arm = 5) {

  r_mat <- matrix(NA_real_, nrow = length(features), ncol = length(arms),
                  dimnames = list(features, arms))
  p_mat <- r_mat

  for (f in features) {

    dat <- df %>%
      select(Sample_ID, chr_arm, TL = all_of(tl_col), Y = all_of(f)) %>%
      filter(complete.cases(.)) %>%
      group_by(chr_arm) %>%
      mutate(n_arm = n()) %>%
      ungroup() %>%
      filter(n_arm >= min_n_per_arm) %>%
      select(-n_arm)

    if (nrow(dat) < 10) next
    if (sd(dat$TL) == 0 || sd(dat$Y) == 0) next

    fit <- tryCatch(
      lmer(Y ~ TL * chr_arm + (1 | Sample_ID), data = dat, REML = FALSE),
      error = function(e) NULL
    )
    if (is.null(fit)) next

    tr <- tryCatch(
      as.data.frame(emmeans::emtrends(fit, ~ chr_arm, var = "TL")),
      error = function(e) NULL
    )
    if (is.null(tr) || nrow(tr) == 0) next

    trend_col <- grep("\\.trend$", names(tr), value = TRUE)
    if (length(trend_col) != 1) next

    for (k in seq_len(nrow(tr))) {
      a <- as.character(tr$chr_arm[k])
      if (is.na(a) || !(a %in% colnames(r_mat))) next

      est <- tr[[trend_col]][k]
      se  <- if ("SE" %in% names(tr)) tr$SE[k] else NA_real_
      if (is.na(est) || is.na(se) || se == 0) next

      tt <- if ("t.ratio" %in% names(tr)) tr$t.ratio[k] else est / se
      if (is.na(tt)) next

      dfk <- if ("df" %in% names(tr) && is.finite(tr$df[k])) tr$df[k] else
        max(nrow(dat) - length(lme4::fixef(fit)), 1)

      pk <- if ("p.value" %in% names(tr) && is.finite(tr$p.value[k])) tr$p.value[k] else
        2 * stats::pt(-abs(tt), df = dfk)

      rk <- tt / sqrt(tt^2 + dfk)

      r_mat[f, a] <- rk
      p_mat[f, a] <- pk
    }
  }

  # BH-FDR per feature (per row) across arms
  fdr_mat <- t(apply(p_mat, 1, function(p_row) {
    ok <- !is.na(p_row)
    out <- rep(NA_real_, length(p_row))
    if (sum(ok) > 0) out[ok] <- p.adjust(p_row[ok], method = "BH")
    out
  }))
  dimnames(fdr_mat) <- dimnames(p_mat)

  list(r = r_mat, p = p_mat, fdr = fdr_mat)
}


# variable names used inside mixed_corr_mats()
# df        :  merged all_df
# tl_col    : name of the TL column to use as predictor (e.g., "norm_TL" or "final_average_TL_p75_kb")
# features  : character vector of feature column names to test (e.g., sv_rate_*, cnv_rate_*, etc.)
# arms      : vector of chr_arm levels (heatmap columns), e.g., "chr1p", "chr1q", ...
# min_n_per_arm : minimum number of non-missing rows required within an arm to analyze that arm for a given feature
#
# r_mat     : output matrix of r-like effect sizes (feature × arm); derived from the TL slope t-statistic
# p_mat     : output matrix of raw p-values (feature × arm) for the TL slope within each arm
#
# dat       : per-feature analysis table after selecting columns, dropping NAs, and filtering small arms
# fit       : fitted mixed-effects model object from lmer()
# tr        : table of per-arm TL slopes returned by emtrends()
#
# trend_col : the column name in 'tr' that contains the TL slope estimate (usually "TL.trend")
# est       : estimated TL slope for a specific chr_arm (change in Y per 1 unit change in TL)
# se        : standard error of the slope estimate
# tt        : t-statistic for the slope (either from emtrends as t.ratio, or computed as est/se)
# dfk       : degrees of freedom for the t-test (prefer emtrends' df; otherwise a rough fallback)
# pk        : raw p-value for the TL slope in that arm (prefer emtrends' p.value; otherwise computed from tt and dfk)
# rk        : r-like signed effect size computed from t and df: r = t / sqrt(t^2 + df)
#
# f         : current feature name in the outer loop
# a         : current chr_arm name in the inner loop
# k         : row index over arms within the emtrends() output



```

```r
res_abs  <- mixed_corr_mats(all_df, "final_average_TL_p75_kb", features, arms, min_n_per_arm = 5)
res_norm <- mixed_corr_mats(all_df, "norm_TL",                 features, arms, min_n_per_arm = 5)

# correlation values based on mixed model, metrices for ComplexHeatmap:
r_abs   <- res_abs$r
fdr_abs <- res_abs$fdr

r_norm   <- res_norm$r
fdr_norm <- res_norm$fdr
```

```r
#── helper for star annotation ─────────────────────────────────────────────────
star <- function(p) {
  if(is.na(p)) return("")
  else if(p < 0.001) return("***")
  else if(p < 0.01)  return("**")
  else if(p < 0.05)  return("*")
  else return("")
}
options(repr.plot.width = 12, repr.plot.height = 3, repr.plot.res = 200)
```

```r
#heatmap with stars representing FDR
```

```r
ht2 <- Heatmap(
  r_norm,
  name            = "r_abs",
  col             = colorRamp2(c(-0.5,0,0.5), c("blue","white","red")),
  cluster_rows    = FALSE,
  cluster_columns = FALSE,
  column_title    = "Absolute TL",
  row_title       = "Feature",
  cell_fun = function(j,i,x,y,w,h,fill) {
    s <- star(fdr_abs[i,j])
    if(nzchar(s)) {
      grid.text(s, x=x, y=y, gp=gpar(fontsize=12))
    }
  },
  heatmap_legend_param = list(title="Mixed model")
)
draw(ht2, heatmap_legend_side="right")
```

```r
mixed_corr_mats <- function(df, tl_col, features, arms, min_n_per_arm = 5) {

  r_mat <- matrix(NA_real_, nrow = length(features), ncol = length(arms),
                  dimnames = list(features, arms))
  p_mat <- r_mat

  for (f in features) {

    dat <- df %>%
      select(Sample_ID, chr_arm, TL = all_of(tl_col), Y = all_of(f)) %>%
      filter(complete.cases(.)) %>%
      group_by(chr_arm) %>%
      mutate(n_arm = n()) %>%
      ungroup() %>%
      filter(n_arm >= min_n_per_arm) %>%
      select(-n_arm)

    # skip if too small / no variation
    if (nrow(dat) < 10) next
    if (sd(dat$TL) == 0 || sd(dat$Y) == 0) next

    # mixed model
    fit <- tryCatch(
      lmer(Y ~ TL * chr_arm + (1 | Sample_ID), data = dat, REML = FALSE),
      error = function(e) NULL
    )
    if (is.null(fit)) next

    # per-arm TL slope
    tr <- tryCatch(
      as.data.frame(emmeans::emtrends(fit, ~ chr_arm, var = "TL")),
      error = function(e) NULL
    )
    if (is.null(tr) || nrow(tr) == 0) next

    # find the slope column (e.g., "TL.trend")
    trend_col <- grep("\\.trend$", names(tr), value = TRUE)
    if (length(trend_col) != 1) next

    for (k in seq_len(nrow(tr))) {
      a <- as.character(tr$chr_arm[k])
      if (is.na(a) || !(a %in% colnames(r_mat))) next

      est <- tr[[trend_col]][k]
      se  <- if ("SE" %in% names(tr)) tr$SE[k] else NA_real_
      if (is.na(est) || is.na(se) || se == 0) next

      # t-stat (use provided if present; otherwise compute)
      tt <- if ("t.ratio" %in% names(tr)) tr$t.ratio[k] else est / se
      if (is.na(tt)) next

      # df (prefer emtrends df; otherwise rough fallback)
      dfk <- if ("df" %in% names(tr) && is.finite(tr$df[k])) tr$df[k] else
        max(nrow(dat) - length(lme4::fixef(fit)), 1)

      # RAW p-value (prefer emtrends p.value; otherwise compute)
      pk <- if ("p.value" %in% names(tr) && is.finite(tr$p.value[k])) tr$p.value[k] else
        2 * stats::pt(-abs(tt), df = dfk)

      # signed r-like effect size from t-stat
      rk <- tt / sqrt(tt^2 + dfk)

      r_mat[f, a] <- rk
      p_mat[f, a] <- pk
    }
  }

  # NO FDR: return raw p-values
  list(r = r_mat, p = p_mat)
}
```

```r
res_abs  <- mixed_corr_mats(all_df, "final_average_TL_p75_kb", features, arms, min_n_per_arm = 5)
res_norm <- mixed_corr_mats(all_df, "norm_TL",                 features, arms, min_n_per_arm = 5)

```

```r
r_abs <- res_abs$r
p_abs <- res_abs$p

r_norm <- res_norm$r
p_norm <- res_norm$p
```
