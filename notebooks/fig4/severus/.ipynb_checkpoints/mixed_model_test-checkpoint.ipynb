```r
#this is a test code to find correlates of telomere lenght with genomic features based on mixed stat models
```

```r
#telomere length is fixed variable

Per-arm detail (heatmap): allow TL effects to vary by arm
Y ~ TL * chr_arm + (1 | Sample_ID) → arm-specific TL slopes (heatmap cells)

Overall effect (forest): one TL effect per feature, pooled across arms
Y ~ TL + chr_arm + (1 | Sample_ID) → one slope ± CI per feature (forest)
```

```r
# Mixed-effects model overview (why we use it here)
# We have repeated measurements per patient/sample (many chr_arms per Sample_ID), so observations are not independent.
# A mixed model accounts for this by modeling Sample_ID as a random effect (patient-specific baseline), while estimating the fixed TL effect (and allowing TL slopes to differ by chr_arm via TL * chr_arm).
# We then extract the per-arm TL slope and its p-value for each genomic feature to build matrices for heatmap visualization (either raw p-values for exploration or BH-FDR for multiple-testing control).
```

```r
library(dplyr)
library(lme4)
library(lmerTest)  # IMPORTANT: load this AFTER lme4
library(emmeans)
```

```r
#This function is testing for the mixed model, calculating FDR

# What mixed_corr_mats() does
# 1) For each genomic feature column, it builds a clean analysis table with TL (chosen column), the feature (Y), chr_arm, and Sample_ID, dropping missing values and optionally filtering arms with small n.
# 2) It fits a mixed model: Y ~ TL * chr_arm + (1 | Sample_ID), which estimates arm-specific TL slopes while accounting for non-independence of arms within the same Sample_ID.
# 3) It uses emtrends() to extract the TL slope per chr_arm, converts the slope t-statistic to an r-like signed effect size, and returns two matrices: r (effect size) and p (raw p-values) aligned as feature × chr_arm for heatmap plotting.



mixed_corr_mats <- function(df, tl_col, features, arms, min_n_per_arm = 5) {

  r_mat <- matrix(NA_real_, nrow = length(features), ncol = length(arms),
                  dimnames = list(features, arms))
  p_mat <- r_mat

  for (f in features) {

    dat <- df %>%
      select(Sample_ID, chr_arm, TL = all_of(tl_col), Y = all_of(f)) %>%
      filter(complete.cases(.)) %>%
      group_by(chr_arm) %>%
      mutate(n_arm = n()) %>%
      ungroup() %>%
      filter(n_arm >= min_n_per_arm) %>%
      select(-n_arm)

    if (nrow(dat) < 10) next
    if (sd(dat$TL) == 0 || sd(dat$Y) == 0) next

    fit <- tryCatch(
      lmer(Y ~ TL * chr_arm + (1 | Sample_ID), data = dat, REML = FALSE),
      error = function(e) NULL
    )
    if (is.null(fit)) next

    tr <- tryCatch(
      as.data.frame(emmeans::emtrends(fit, ~ chr_arm, var = "TL")),
      error = function(e) NULL
    )
    if (is.null(tr) || nrow(tr) == 0) next

    trend_col <- grep("\\.trend$", names(tr), value = TRUE)
    if (length(trend_col) != 1) next

    for (k in seq_len(nrow(tr))) {
      a <- as.character(tr$chr_arm[k])
      if (is.na(a) || !(a %in% colnames(r_mat))) next

      est <- tr[[trend_col]][k]
      se  <- if ("SE" %in% names(tr)) tr$SE[k] else NA_real_
      if (is.na(est) || is.na(se) || se == 0) next

      tt <- if ("t.ratio" %in% names(tr)) tr$t.ratio[k] else est / se
      if (is.na(tt)) next

      dfk <- if ("df" %in% names(tr) && is.finite(tr$df[k])) tr$df[k] else
        max(nrow(dat) - length(lme4::fixef(fit)), 1)

      pk <- if ("p.value" %in% names(tr) && is.finite(tr$p.value[k])) tr$p.value[k] else
        2 * stats::pt(-abs(tt), df = dfk)

      rk <- tt / sqrt(tt^2 + dfk)

      r_mat[f, a] <- rk
      p_mat[f, a] <- pk
    }
  }

  # BH-FDR per feature (per row) across arms
  fdr_mat <- t(apply(p_mat, 1, function(p_row) {
    ok <- !is.na(p_row)
    out <- rep(NA_real_, length(p_row))
    if (sum(ok) > 0) out[ok] <- p.adjust(p_row[ok], method = "BH")
    out
  }))
  dimnames(fdr_mat) <- dimnames(p_mat)

  list(r = r_mat, p = p_mat, fdr = fdr_mat)
}


# variable names used inside mixed_corr_mats()
# df        :  merged all_df
# tl_col    : name of the TL column to use as predictor (e.g., "norm_TL" or "final_average_TL_p75_kb")
# features  : character vector of feature column names to test (e.g., sv_rate_*, cnv_rate_*, etc.)
# arms      : vector of chr_arm levels (heatmap columns), e.g., "chr1p", "chr1q", ...
# min_n_per_arm : minimum number of non-missing rows required within an arm to analyze that arm for a given feature
#
# r_mat     : output matrix of r-like effect sizes (feature × arm); derived from the TL slope t-statistic
# p_mat     : output matrix of raw p-values (feature × arm) for the TL slope within each arm
#
# dat       : per-feature analysis table after selecting columns, dropping NAs, and filtering small arms
# fit       : fitted mixed-effects model object from lmer()
# tr        : table of per-arm TL slopes returned by emtrends()
#
# trend_col : the column name in 'tr' that contains the TL slope estimate (usually "TL.trend")
# est       : estimated TL slope for a specific chr_arm (change in Y per 1 unit change in TL)
# se        : standard error of the slope estimate
# tt        : t-statistic for the slope (either from emtrends as t.ratio, or computed as est/se)
# dfk       : degrees of freedom for the t-test (prefer emtrends' df; otherwise a rough fallback)
# pk        : raw p-value for the TL slope in that arm (prefer emtrends' p.value; otherwise computed from tt and dfk)
# rk        : r-like signed effect size computed from t and df: r = t / sqrt(t^2 + df)
#
# f         : current feature name in the outer loop
# a         : current chr_arm name in the inner loop
# k         : row index over arms within the emtrends() output



```

```r
res_abs  <- mixed_corr_mats(all_df, "final_average_TL_p75_kb", features, arms, min_n_per_arm = 5)
res_norm <- mixed_corr_mats(all_df, "norm_TL",                 features, arms, min_n_per_arm = 5)

# correlation values based on mixed model, metrices for ComplexHeatmap:
r_abs   <- res_abs$r
fdr_abs <- res_abs$fdr

r_norm   <- res_norm$r
fdr_norm <- res_norm$fdr
```

```r
###############################################################################
# Mixed model (per feature, pooled across ALL arms):
#   Y ~ TL + chr_arm + (1 | Sample_ID)
# - TL is the main predictor
# - chr_arm adjusts arm-specific baseline differences
# - (1|Sample_ID) accounts for repeated arms within each patient
#
###############################################################################

# ── 0) Libraries ─────────────────────────────────────────────────────────────
library(dplyr)
library(lme4)
library(lmerTest)   # after lme4
library(ggplot2)

# ── 1) Inputs ────────────────────────────────────────────────────────────────
features <- grep("^(sv_rate|sv_proportion|cnv_rate_per_mb|cnv_proportion_per_mb)_",
                 names(all_df), value = TRUE)

all_df <- all_df %>%
  mutate(
    Sample_ID = factor(Sample_ID),
    chr_arm   = factor(chr_arm)
  )

# ── 2) Settings ──────────────────────────────────────────────────────────────
TOP_N          <- 15
MIN_N_TOTAL    <- 50     # minimum complete rows required per feature
STANDARDIZE_Y  <- TRUE   # z-score Y per feature so slopes are comparable
PLOT_WIDTH     <- 10
PLOT_HEIGHT    <- 4
PLOT_RES       <- 300

options(repr.plot.width = PLOT_WIDTH,
        repr.plot.height = PLOT_HEIGHT,
        repr.plot.res = PLOT_RES)

# ── 3) Fit one pooled mixed model per feature and extract TL effect ───────────
get_overall_TL_effects_rawp <- function(df, tl_col, features,
                                       min_n_total = 50,
                                       standardize_y = TRUE) {

  out <- vector("list", length(features))

  for (i in seq_along(features)) {
    f <- features[i]

    dat <- df %>%
      transmute(
        Sample_ID = factor(Sample_ID),
        chr_arm   = factor(chr_arm),
        TL        = .data[[tl_col]],
        Y         = .data[[f]]
      ) %>%
      filter(complete.cases(.))

    if (nrow(dat) < min_n_total) next
    if (sd(dat$TL) == 0 || sd(dat$Y) == 0) next

    if (standardize_y) dat$Y <- as.numeric(scale(dat$Y))

    fit <- tryCatch(
     lmer(Y ~ TL + chr_arm + (1 | Sample_ID), data = dat, REML = FALSE),

      error = function(e) NULL
    )
    if (is.null(fit)) next

    co <- summary(fit)$coefficients
    if (!("TL" %in% rownames(co))) next

    beta <- co["TL", "Estimate"]
    se   <- co["TL", "Std. Error"]
    dfk  <- co["TL", "df"]
    pval <- co["TL", "Pr(>|t|)"]

    tcrit <- qt(0.975, df = dfk)
    lo <- beta - tcrit * se
    hi <- beta + tcrit * se

    out[[i]] <- data.frame(
      feature  = f,
      estimate = beta,
      lower    = lo,
      upper    = hi,
      p        = pval,
      n        = nrow(dat),
      stringsAsFactors = FALSE
    )
  }

  bind_rows(out) %>% arrange(p)
}

# ── 4) Plot function (raw p-values; shows direction by sign) ──────────────────
plot_forest_features_rawp <- function(df_eff, title, top_n = 15) {

  df_top <- df_eff %>%
    filter(!is.na(estimate), !is.na(lower), !is.na(upper), !is.na(p)) %>%
    arrange(p) %>%
    slice_head(n = top_n) %>%
    mutate(
      feature = factor(feature, levels = rev(feature)),
      direction = ifelse(estimate > 0, "Positive (TL↑ → feature↑)", "Negative (TL↑ → feature↓)"),
      p_label = sprintf("p=%.2g", p)
    )

  ggplot(df_top, aes(x = estimate, y = feature)) +
    geom_vline(xintercept = 0, linetype = "dashed", linewidth = 0.4) +
    geom_pointrange(aes(xmin = lower, xmax = upper, shape = direction),
                    linewidth = 0.6) +
    #geom_text(aes(label = p_label), hjust = -0.10, size = 3.2) +
    coord_cartesian(clip = "off") +
    labs(
      title = title,
      x = "Effect of telomere length on feature (slope per 1 unit TL) ± 95% CI",
      y = NULL,
      shape = "Direction"
    ) +
    theme_classic(base_size = 12) +
    theme(
      plot.title  = element_text(face = "bold", size = 13),
      axis.title.x = element_text(face = "bold"),
      axis.text.y = element_text(size = 10),
      plot.margin = margin(10, 95, 10, 10)
    )
}

# ── 5) Compute effects (Normalized TL + Absolute TL) ──────────────────────────
eff_norm <- get_overall_TL_effects_rawp(
  all_df, tl_col = "norm_TL", features = features,
  min_n_total = MIN_N_TOTAL, standardize_y = STANDARDIZE_Y
)

eff_abs <- get_overall_TL_effects_rawp(
  all_df, tl_col = "final_average_TL_p75_kb", features = features,
  min_n_total = MIN_N_TOTAL, standardize_y = STANDARDIZE_Y
)

# ── 6) Make forest plots (TOP_N features by raw p) ────────────────────────────
suffix <- if (STANDARDIZE_Y) " (Y z-scored per feature)" else " (native units)"

p_norm <- plot_forest_features_rawp(
  eff_norm,
  title = paste0("Top ", TOP_N, " features vs Normalized TL", suffix),
  top_n = TOP_N
)

p_abs <- plot_forest_features_rawp(
  eff_abs,
  title = paste0("Top ", TOP_N, " features vs Absolute TL", suffix),
  top_n = TOP_N
)

print(p_norm)
print(p_abs)
```

```r
######pateint fixed here##############
# ── 0) Libraries ─────────────────────────────────────────────────────────────
library(dplyr)
library(ggplot2)

# ── 1) Inputs ────────────────────────────────────────────────────────────────
features <- grep("^(sv_rate|sv_proportion|cnv_rate_per_mb|cnv_proportion_per_mb)_",
                 names(all_df), value = TRUE)

all_df <- all_df %>%
  mutate(
    Sample_ID = factor(Sample_ID),
    chr_arm   = factor(chr_arm)
  )

# ── 2) Settings ──────────────────────────────────────────────────────────────
TOP_N         <- 15
MIN_N_TOTAL   <- 50     # minimum complete rows per feature
STANDARDIZE_Y <- TRUE   # z-score Y per feature for comparable slopes
PLOT_WIDTH    <- 10
PLOT_HEIGHT   <- 4
PLOT_RES      <- 300

options(repr.plot.width  = PLOT_WIDTH,
        repr.plot.height = PLOT_HEIGHT,
        repr.plot.res    = PLOT_RES)

# ── 3) Fit one PATIENT-FE model per feature and extract TL effect ─────────────
get_overall_TL_effects_patientFE_rawp <- function(df, tl_col, features,
                                                 min_n_total = 50,
                                                 standardize_y = TRUE) {

  out <- vector("list", length(features))

  for (i in seq_along(features)) {
    f <- features[i]

    dat <- df %>%
      transmute(
        Sample_ID = factor(Sample_ID),
        chr_arm   = factor(chr_arm),
        TL        = .data[[tl_col]],
        Y         = .data[[f]]
      ) %>%
      filter(complete.cases(.))

    if (nrow(dat) < min_n_total) next
    if (sd(dat$TL) == 0 || sd(dat$Y) == 0) next

    if (standardize_y) dat$Y <- as.numeric(scale(dat$Y))

    fit <- tryCatch(
      lm(Y ~ TL + chr_arm + Sample_ID, data = dat),
      error = function(e) NULL
    )
    if (is.null(fit)) next

    co <- summary(fit)$coefficients
    if (!("TL" %in% rownames(co))) next

    beta <- co["TL", "Estimate"]
    se   <- co["TL", "Std. Error"]
    pval <- co["TL", "Pr(>|t|)"]
    dfk  <- df.residual(fit)

    tcrit <- qt(0.975, df = dfk)
    lo <- beta - tcrit * se
    hi <- beta + tcrit * se

    out[[i]] <- data.frame(
      feature  = f,
      estimate = beta,
      lower    = lo,
      upper    = hi,
      p        = pval,
      n        = nrow(dat),
      stringsAsFactors = FALSE
    )
  }

  bind_rows(out) %>% arrange(p)
}

# ── 4) Plot function (raw p-values; explicit direction by sign) ───────────────
plot_forest_features_rawp <- function(df_eff, title, top_n = 15) {

  df_top <- df_eff %>%
    filter(!is.na(estimate), !is.na(lower), !is.na(upper), !is.na(p)) %>%
    arrange(p) %>%
    slice_head(n = top_n) %>%
    mutate(
      feature = factor(feature, levels = rev(feature)),
      direction = ifelse(estimate > 0, "Positive (TL↑ → feature↑)", "Negative (TL↑ → feature↓)"),
      p_label = sprintf("p=%.2g", p)
    )

  ggplot(df_top, aes(x = estimate, y = feature)) +
    geom_vline(xintercept = 0, linetype = "dashed", linewidth = 0.4) +
    geom_pointrange(aes(xmin = lower, xmax = upper, shape = direction),
                    linewidth = 0.6) +
    #geom_text(aes(label = p_label), hjust = -0.10, size = 3.2) +
    coord_cartesian(clip = "off") +
    labs(
      title = title,
      x = "Effect of telomere length on feature (slope per 1 unit TL) ± 95% CI",
      y = NULL,
      shape = "Direction"
    ) +
    theme_classic(base_size = 12) +
    theme(
      plot.title   = element_text(face = "bold", size = 13),
      axis.title.x = element_text(face = "bold"),
      axis.text.y  = element_text(size = 10),
      plot.margin  = margin(10, 95, 10, 10)
    )
}

# ── 5) Compute effects (Normalized TL + Absolute TL) ──────────────────────────
eff_norm <- get_overall_TL_effects_patientFE_rawp(
  all_df, tl_col = "norm_TL", features = features,
  min_n_total = MIN_N_TOTAL, standardize_y = STANDARDIZE_Y
)

eff_abs <- get_overall_TL_effects_patientFE_rawp(
  all_df, tl_col = "final_average_TL_p75_kb", features = features,
  min_n_total = MIN_N_TOTAL, standardize_y = STANDARDIZE_Y
)

# ── 6) Forest plots (TOP_N features by raw p) ─────────────────────────────────
suffix <- if (STANDARDIZE_Y) " (Y z-scored per feature)" else " (native units)"

p_norm <- plot_forest_features_rawp(
  eff_norm,
  title = paste0("Top ", TOP_N, " features vs Normalized TL (patient fixed effects)", suffix),
  top_n = TOP_N
)

p_abs <- plot_forest_features_rawp(
  eff_abs,
  title = paste0("Top ", TOP_N, " features vs Absolute TL (patient fixed effects)", suffix),
  top_n = TOP_N
)

print(p_norm)
print(p_abs)

```

```r
# ── 0) Libraries ─────────────────────────────────────────────────────────────
library(dplyr)
library(lme4)
library(lmerTest)   # p-values, df
library(ggplot2)

# ── 1) Inputs ────────────────────────────────────────────────────────────────
# SV/CNV-derived features to test
features <- grep(
  "^(sv_rate|sv_proportion|cnv_rate_per_mb|cnv_proportion_per_mb)_",
  names(all_df),
  value = TRUE
)

all_df <- all_df %>%
  mutate(
    Sample_ID = factor(Sample_ID),
    chr_arm   = factor(chr_arm)
  )

# ── 2) Settings ──────────────────────────────────────────────────────────────
TOP_N          <- 15
MIN_N_TOTAL    <- 50     # minimum complete rows per feature
STANDARDIZE_SV <- TRUE  # z-score SV so slopes are comparable
PLOT_WIDTH     <- 10
PLOT_HEIGHT    <- 4
PLOT_RES       <- 300

options(
  repr.plot.width  = PLOT_WIDTH,
  repr.plot.height = PLOT_HEIGHT,
  repr.plot.res    = PLOT_RES
)

# ── 3) Fit pooled mixed model per feature: TL ~ SV + chr_arm + (1|Sample_ID) ──
get_overall_SV_effects_on_TL_rawp <- function(df, tl_col, features,
                                             min_n_total = 50,
                                             standardize_sv = TRUE) {

  out <- vector("list", length(features))

  for (i in seq_along(features)) {
    f <- features[i]

    dat <- df %>%
      transmute(
        Sample_ID = factor(Sample_ID),
        chr_arm   = factor(chr_arm),
        TL        = .data[[tl_col]],
        SV        = .data[[f]]
      ) %>%
      filter(complete.cases(.))

    if (nrow(dat) < min_n_total) next
    if (sd(dat$TL) == 0 || sd(dat$SV) == 0) next

    # Standardize predictor (SV), not outcome
    if (standardize_sv) dat$SV <- as.numeric(scale(dat$SV))

    fit <- tryCatch(
      lmer(
        TL ~ SV + chr_arm + (1 | Sample_ID),
        data = dat,
        REML = FALSE
      ),
      error = function(e) NULL
    )
    if (is.null(fit)) next

    co <- summary(fit)$coefficients
    if (!("SV" %in% rownames(co))) next

    beta <- co["SV", "Estimate"]
    se   <- co["SV", "Std. Error"]
    dfk  <- co["SV", "df"]
    pval <- co["SV", "Pr(>|t|)"]

    tcrit <- qt(0.975, df = dfk)
    lo <- beta - tcrit * se
    hi <- beta + tcrit * se

    out[[i]] <- data.frame(
      feature  = f,
      estimate = beta,
      lower    = lo,
      upper    = hi,
      p        = pval,
      n        = nrow(dat),
      stringsAsFactors = FALSE
    )
  }

  bind_rows(out) %>% arrange(p)
}

# ── 4) Forest plot function (raw p-values; SV → TL direction) ─────────────────
plot_forest_features_rawp <- function(df_eff, title, top_n = 15) {

  df_top <- df_eff %>%
    filter(!is.na(estimate), !is.na(lower), !is.na(upper), !is.na(p)) %>%
    arrange(p) %>%
    slice_head(n = top_n) %>%
    mutate(
      feature = factor(feature, levels = rev(feature)),
      direction = ifelse(
        estimate > 0,
        "More SVs → longer TL",
        "More SVs → shorter TL"
      )
    )

  ggplot(df_top, aes(x = estimate, y = feature)) +
    geom_vline(xintercept = 0, linetype = "dashed", linewidth = 0.4) +
    geom_pointrange(
      aes(xmin = lower, xmax = upper, shape = direction),
      linewidth = 0.6
    ) +
    coord_cartesian(clip = "off") +
    labs(
      title = title,
      x = "Effect of SV burden on telomere length (ΔTL per 1 SD SV) ± 95% CI",
      y = NULL,
      shape = "Direction"
    ) +
    theme_classic(base_size = 12) +
    theme(
      plot.title   = element_text(face = "bold", size = 13),
      axis.title.x = element_text(face = "bold"),
      axis.text.y  = element_text(size = 10),
      plot.margin  = margin(10, 95, 10, 10)
    )
}

# ── 5) Compute effects (Normalized TL and Absolute TL) ─────────────────────────
eff_norm <- get_overall_SV_effects_on_TL_rawp(
  all_df,
  tl_col = "norm_TL",
  features = features,
  min_n_total = MIN_N_TOTAL,
  standardize_sv = STANDARDIZE_SV
)

eff_abs <- get_overall_SV_effects_on_TL_rawp(
  all_df,
  tl_col = "final_average_TL_p75_kb",
  features = features,
  min_n_total = MIN_N_TOTAL,
  standardize_sv = STANDARDIZE_SV
)

# ── 6) Forest plots (TOP_N by raw p-value) ─────────────────────────────────────
suffix <- if (STANDARDIZE_SV) " (SV z-scored)" else " (native units)"

p_norm <- plot_forest_features_rawp(
  eff_norm,
  title = paste0("Top ", TOP_N, " SV/CNV features → Normalized TL", suffix),
  top_n = TOP_N
)

p_abs <- plot_forest_features_rawp(
  eff_abs,
  title = paste0("Top ", TOP_N, " SV/CNV features → Absolute TL", suffix),
  top_n = TOP_N
)

print(p_norm)
print(p_abs)
```
