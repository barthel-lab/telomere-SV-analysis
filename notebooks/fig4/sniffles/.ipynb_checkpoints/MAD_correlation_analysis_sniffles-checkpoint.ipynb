```r
library(dplyr)
library(tidyr)
library(ComplexHeatmap)
library(circlize)
library(grid)

# Prepare telomere data: use norm_TL as RTL
telomere_long <- telomeres_MAD %>%
  rename(Sample_ID = Sample_prefix) %>%
  select(Sample_ID, chr_arm, RTL = norm_TL) %>%
  filter(!is.na(RTL))

# Prepare CNV data
cnv_sel <- cnv %>%
  rename(Sample_ID = ID, chr_arm = arm) %>%
  filter(Gain_Loss %in% c("Gain", "Loss")) %>%
  select(Sample_ID, chr_arm, Gain_Loss, cnv_rate_per_mb, cnv_proportion_per_mb)

# Prepare SV data
sv_sel <- sv %>%
  filter(type %in% c("DEL", "DUP", "INV", "INS")) %>%
  select(type, chr_arm, Sample_ID, sv_rate, sv_proportion)

# Safe correlation function
#safe_cor <- function(x, y) {
 # if (sum(complete.cases(x, y)) > 1) cor(x, y, use = "complete.obs") else NA_real_
#}

safe_cor <- function(x, y, method = "pearson") {
  if (sum(complete.cases(x, y)) > 1) {
    cor(x, y, use = "complete.obs", method = method)
  } else {
    NA_real_
  }
}
# Correlation calculation function
cor_by_arm <- function(df, value_col, var_prefix) {
  df_joined <- df %>%
    inner_join(telomere_long, by = c("Sample_ID", "chr_arm")) %>%
    filter(!is.na(RTL))
  
  group_vars <- if ("type" %in% colnames(df_joined)) {
    c("type", "chr_arm")
  } else if ("Gain_Loss" %in% colnames(df_joined)) {
    c("Gain_Loss", "chr_arm")
  } else {
    "chr_arm"
  }
  
  df_joined <- df_joined %>%
    mutate(value = .data[[value_col]])
  
  summarised <- df_joined %>%
    group_by(across(all_of(group_vars))) %>%
    summarise(
      r = safe_cor(value, RTL),
      p = if (sum(complete.cases(value, RTL)) > 2) cor.test(value, RTL)$p.value else NA_real_,
      .groups = "drop"
    )
  
  summarised <- summarised %>%
    mutate(variable = if ("type" %in% group_vars) {
      paste0(var_prefix, "_", type)
    } else if ("Gain_Loss" %in% group_vars) {
      paste0(var_prefix, "_", Gain_Loss)
    } else {
      var_prefix
    })
  
  summarised
}

# Run correlations
cnv_r1 <- cor_by_arm(cnv_sel, "cnv_rate_per_mb", "cnv_rate")
cnv_r2 <- cor_by_arm(cnv_sel, "cnv_proportion_per_mb", "cnv_prop")

sv_r1 <- cor_by_arm(sv_sel, "sv_rate", "sv_rate")
sv_r2 <- cor_by_arm(sv_sel, "sv_proportion", "sv_prop")

cor_df <- bind_rows(cnv_r1, cnv_r2, sv_r1, sv_r2)

# Prepare matrices for heatmap
corr_matrix <- cor_df %>%
  select(variable, chr_arm, r) %>%
  pivot_wider(names_from = chr_arm, values_from = r) %>%
  column_to_rownames("variable") %>%
  as.matrix()

pval_matrix <- cor_df %>%
  select(variable, chr_arm, p) %>%
  pivot_wider(names_from = chr_arm, values_from = p) %>%
  column_to_rownames("variable") %>%
  as.matrix()

# Heatmap colors and significance stars
col_fun <- colorRamp2(c(-1, 0, 1), c("blue", "white", "red"))

pval_to_stars <- function(p) {
  if (is.na(p)) return("")
  else if (p < 0.001) return("***")
  else if (p < 0.01) return("**")
  else if (p < 0.05) return("*")
  else return("")
}

heatmap_p_q <- Heatmap(
  corr_matrix,
  name = "Correlation",
  col = col_fun,
  na_col = "grey",
  rect_gp = gpar(col = "black", lwd = 0.5),
  cluster_rows = TRUE,
  cluster_columns = FALSE,
  column_title = "Telomere length (MAD normalized) vs Variant features",
  row_title = "Variant features",
  heatmap_legend_param = list(title = "r"),
  cell_fun = function(j, i, x, y, width, height, fill) {
    pval <- pval_matrix[i, j]
    if (!is.na(pval)) {
      r_val <- corr_matrix[i, j]
      stars <- pval_to_stars(pval)
      is_sig <- nzchar(stars)
      
      fs <- if (is_sig) 15 else 12
      fface <- if (is_sig) "bold" else "plain"
      
      log_p <- -log10(pval)
      alpha <- min(log_p / 3, 1)
      text_a <- 0.1 + 0.7 * alpha
      text_col <- rgb(0, 0, 0, text_a)
      
      grid.text(sprintf("%.2f%s", r_val, stars), x, y, gp = gpar(fontsize = fs, fontface = fface, col = text_col))
    }
  }
)

p_legend <- Legend(
  title = "Significance",
  labels = c("*** p<0.001", "** p<0.01", "* p<0.05"),
  legend_gp = gpar(fill = NA),
  type = "lines"
)

draw(heatmap_p_q, annotation_legend_list = list(p_legend))

```

```r
library(dplyr)

summary_simple <- cor_df %>%
  filter(!is.na(r)) %>%
  group_by(variable) %>%
  summarise(
    mean_r = mean(r, na.rm = TRUE),                 # average correlation
    sig_pos_count = sum(r > 0 & p < 0.05, na.rm = TRUE),  # count positive significant correlations
    sig_neg_count = sum(r < 0 & p < 0.05, na.rm = TRUE),  # count negative significant correlations
    total_arms = n()                                # total number of arms / data points
  ) %>%
  ungroup() %>%
  mutate(
    direction = case_when(
      mean_r > 0 ~ "Positive",
      mean_r < 0 ~ "Negative",
      TRUE ~ "Neutral"
    )
  )
```

```r
# Full script: per-arm correlations & two heatmaps with significance stars

#── 0. Load packages ───────────────────────────────────────────────────────────
library(dplyr)
library(tidyr)
library(ComplexHeatmap)
library(circlize)
library(grid)

#── 1. SV: wide by type ────────────────────────────────────────────────────────
sv_wide <- sv %>%
  mutate(chr_arm = paste0(chr, arm)) %>%
  select(Sample_ID, chr_arm, type, sv_rate, sv_proportion) %>%
  pivot_wider(
    names_from  = type,
    values_from = c(sv_rate, sv_proportion),
    values_fill = list(sv_rate = 0, sv_proportion = 0)
  ) %>%
  rename_with(~ gsub(" ", "", .x))

#── 2. CNV: recode & aggregate ─────────────────────────────────────────────────
cnv_summarized <- cnv %>%
  mutate(
    chr_arm   = paste0(chrom, Arm),
    rate_Gain = ifelse(Gain_Loss=="Gain", cnv_rate_per_mb, 0),
    rate_Loss = ifelse(Gain_Loss=="Loss", cnv_rate_per_mb, 0),
    prop_Gain = ifelse(Gain_Loss=="Gain", cnv_proportion_per_mb, 0),
    prop_Loss = ifelse(Gain_Loss=="Loss", cnv_proportion_per_mb, 0)
  ) %>%
  group_by(ID, chr_arm) %>%
  summarise(
    cnv_rate_per_mb_Gain       = sum(rate_Gain, na.rm=TRUE),
    cnv_rate_per_mb_Loss       = sum(rate_Loss, na.rm=TRUE),
    cnv_proportion_per_mb_Gain = sum(prop_Gain, na.rm=TRUE),
    cnv_proportion_per_mb_Loss = sum(prop_Loss, na.rm=TRUE),
    .groups = "drop"
  ) %>%
  rename(Sample_ID = ID)

#── 3. Telomeres ───────────────────────────────────────────────────────────────
tel <- telomeres_MAD %>%
  rename(Sample_ID = Sample_prefix) %>%
  select(Sample_ID, chr_arm, final_average_TL_p75_kb, norm_TL)

#── 4. Merge ──────────────────────────────────────────────────────────────────
all_df <- tel %>%
  left_join(sv_wide, by=c("Sample_ID","chr_arm")) %>%
  left_join(cnv_summarized, by=c("Sample_ID","chr_arm"))

#── 5. Features & arms ────────────────────────────────────────────────────────
features <- grep("^(sv_rate|sv_proportion|cnv_rate_per_mb|cnv_proportion_per_mb)_",
                 names(all_df), value=TRUE)
arms    <- sort(unique(all_df$chr_arm))

#── 6. Compute corr & p matrices safely ───────────────────────────────────────
corr_abs  <- matrix(NA, nrow=length(features), ncol=length(arms),
                    dimnames=list(features, arms))
corr_norm <- corr_abs
p_abs     <- corr_abs
p_norm    <- corr_abs

for(a in arms) {
  sub <- filter(all_df, chr_arm==a)
  for(f in features) {
    x_abs  <- sub[[f]];    y_abs  <- sub$final_average_TL_p75_kb
    x_norm <- sub[[f]];    y_norm <- sub$norm_TL

    ok_abs  <- complete.cases(x_abs,  y_abs)
    ok_norm <- complete.cases(x_norm, y_norm)

    if(sum(ok_abs) >= 3) {
      #t <- cor.test(x_abs[ok_abs],  y_abs[ok_abs])
        t <- cor.test(x_abs[ok_abs],  y_abs[ok_abs], method = "pearson")

      corr_abs[f,a] <- t$estimate; p_abs[f,a]  <- t$p.value
    }
    if(sum(ok_norm) >= 3) {
      #t2 <- cor.test(x_norm[ok_norm], y_norm[ok_norm])
        t2 <- cor.test(x_norm[ok_norm], y_norm[ok_norm], method = "pearson")

      corr_norm[f,a]<- t2$estimate; p_norm[f,a] <- t2$p.value
    }
  }
}


#── helper for star annotation ─────────────────────────────────────────────────
star <- function(p) {
  if(is.na(p)) return("")
  else if(p < 0.001) return("***")
  else if(p < 0.01)  return("**")
  else if(p < 0.05)  return("*")
  else return("")
}
options(repr.plot.width = 15, repr.plot.height = 5, repr.plot.res = 200)

    
#── 7. Plot: Absolute TL ───────────────────────────────────────────────────────
ht1 <- Heatmap(
  corr_abs,
  name            = "r_abs",
  col             = colorRamp2(c(-1,0,1), c("blue","white","red")),
  cluster_rows    = TRUE,
  cluster_columns = TRUE,
  column_title    = "Absolute TL",
  row_title       = "Feature",
  cell_fun = function(j,i,x,y,w,h,fill) {
    s <- star(p_abs[i,j])
    if(nzchar(s)) {
      grid.text(s, x=x, y=y, gp=gpar(fontsize=12))
    }
  },
  heatmap_legend_param = list(title="Pearson r")
)
draw(ht1, heatmap_legend_side="right")

#── 8. Plot: Normalized TL ────────────────────────────────────────────────────
grid.newpage()
ht2 <- Heatmap(
  corr_norm,
  name            = "r_norm",
  col             = colorRamp2(c(-1,0,1), c("blue","white","red")),
  cluster_rows    = TRUE,
  cluster_columns = TRUE,
  column_title    = "Normalized TL",
  row_title       = "Feature",
  cell_fun = function(j,i,x,y,w,h,fill) {
    s <- star(p_norm[i,j])
    if(nzchar(s)) {
      grid.text(s, x=x, y=y, gp=gpar(fontsize=12))
    }
  },
  heatmap_legend_param = list(title="Pearson r")
)
draw(ht2, heatmap_legend_side="right")
```

```r
library(dplyr)
library(ggplot2)

# Prepare data frame of all combinations (features x chromosome arms)
volcano_df <- expand.grid(
  feature = rownames(corr_abs),
  chr_arm = colnames(corr_abs),
  stringsAsFactors = FALSE
)

# Extract correlation and p-value into the data frame
volcano_df <- volcano_df %>%
  mutate(
    corr = as.vector(corr_abs),
    pval = as.vector(p_abs),
    neglog10p = -log10(pval)
  ) %>%
  filter(!is.na(corr) & !is.na(pval))  # remove missing values

# Add significance category
volcano_df <- volcano_df %>%
  mutate(
    sig = case_when(
      pval < 0.001 ~ "***",
      pval < 0.01  ~ "**",
      pval < 0.05  ~ "*",
      TRUE         ~ "ns"
    ),
    sig_flag = pval < 0.05
  )
options(repr.plot.width = 12, repr.plot.height = 4, repr.plot.res = 200)

# Volcano plot
ggplot(volcano_df, aes(x = corr, y = neglog10p)) +
  geom_point(aes(color = sig_flag), alpha = 0.7) +
  geom_text(data = subset(volcano_df, sig_flag),
            aes(label = paste0(feature, "\n", chr_arm)),
            size = 3, vjust = 1.5, check_overlap = TRUE) +
  scale_color_manual(values = c("FALSE" = "grey70", "TRUE" = "red")) +
  labs(
    title = "Volcano plot of correlation vs significance",
    x = "Correlation coefficient (r)",
    y = expression(-log[10](p-value)),
    color = "Significant (p < 0.05)"
  ) +
  theme_minimal(base_size = 14)
```

```r
library(dplyr)
library(ggplot2)

# Compute median TL per arm to order arms by median telomere length (absolute)
arm_order_abs <- all_df %>%
  group_by(chr_arm) %>%
  summarize(median_TL = median(final_average_TL_p75_kb, na.rm = TRUE)) %>%
  arrange(desc(median_TL)) %>%
  pull(chr_arm)

# Compute median TL per arm to order arms by median telomere length (normalized)
arm_order_norm <- all_df %>%
  group_by(chr_arm) %>%
  summarize(median_TL = median(norm_TL, na.rm = TRUE)) %>%
  arrange(desc(median_TL)) %>%
  pull(chr_arm)

# Convert chr_arm to factor with levels sorted by median TL (absolute)
all_df <- all_df %>%
  mutate(
    chr_arm_abs = factor(chr_arm, levels = arm_order_abs),
    chr_arm_norm = factor(chr_arm, levels = arm_order_norm)
  )
options(repr.plot.width = 8, repr.plot.height = 4, repr.plot.res = 200)

# Violin plot for absolute TL sorted by median TL descending
p_abs <- ggplot(all_df, aes(x = chr_arm_abs, y = final_average_TL_p75_kb)) +
  geom_violin(fill = "skyblue", alpha = 0.6) +
  geom_boxplot(width = 0.1, outlier.shape = NA) +
  stat_summary(fun = median, geom = "point", color = "red", size = 2) +
  labs(
    title = "Telomere Length (Absolute) Distribution by Chromosome Arm",
    x = "Chromosome Arm (sorted by median TL)",
    y = "Absolute Telomere Length (p75 kb)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Violin plot for normalized TL sorted by median TL descending
p_norm <- ggplot(all_df, aes(x = chr_arm_norm, y = norm_TL)) +
  geom_violin(fill = "forestgreen", alpha = 0.6) +
  geom_boxplot(width = 0.1, outlier.shape = NA) +
  stat_summary(fun = median, geom = "point", color = "red", size = 2) +
  labs(
    title = "Telomere Length (Normalized) Distribution by Chromosome Arm",
    x = "Chromosome Arm (sorted by median TL)",
    y = "Normalized Telomere Length"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Print plots
print(p_abs)
print(p_norm)
```

```r
library(dplyr)
library(ggplot2)
library(tidyr)
library(ggsignif)

# === STEP 1: Telomere deviation groups ===
grand_median_norm <- median(telomeres_MAD$norm_TL, na.rm = TRUE)

arm_deviation <- telomeres_MAD %>%
  group_by(chr_arm) %>%
  summarise(median_TL = median(norm_TL, na.rm = TRUE)) %>%
  mutate(direction = ifelse(median_TL > grand_median_norm, "Long", "Short"))

# === STEP 2: Label SV and CNV data with telomere group ===
sv_labeled <- sv %>%
  left_join(arm_deviation %>% select(chr_arm, direction), by = "chr_arm")

cnv_labeled <- cnv %>%
  rename(chr_arm = arm) %>%
  left_join(arm_deviation %>% select(chr_arm, direction), by = "chr_arm")

# === STEP 3: Summarize SV data per chr_arm, type, direction ===
sv_summary <- sv_labeled %>%
  group_by(chr_arm, type, direction) %>%
  summarise(
    mean_sv_rate = mean(sv_rate, na.rm = TRUE),
    mean_sv_proportion = mean(sv_proportion, na.rm = TRUE),
    .groups = "drop"
  )

# === STEP 4: Summarize CNV data per chr_arm, Gain_Loss, direction ===
cnv_summary <- cnv_labeled %>%
  group_by(chr_arm, Gain_Loss, direction) %>%
  summarise(
    mean_cnv_rate = mean(cnv_rate_per_mb, na.rm = TRUE),
    mean_cnv_proportion = mean(cnv_proportion_per_mb, na.rm = TRUE),
    .groups = "drop"
  )

# === STEP 5: Wilcoxon test p-values for SV rate ===
sv_pvals <- sv_summary %>%
  group_by(type) %>%
  summarise(p_value = wilcox.test(mean_sv_rate ~ direction)$p.value) %>%
  mutate(signif = case_when(
    p_value < 0.001 ~ "***",
    p_value < 0.01 ~ "**",
    p_value < 0.05 ~ "*",
    TRUE ~ "ns"
  ))

# === STEP 6: Wilcoxon test p-values for CNV rate ===
cnv_pvals <- cnv_summary %>%
  group_by(Gain_Loss) %>%
  summarise(p_value = wilcox.test(mean_cnv_rate ~ direction)$p.value) %>%
  mutate(signif = case_when(
    p_value < 0.001 ~ "***",
    p_value < 0.01 ~ "**",
    p_value < 0.05 ~ "*",
    TRUE ~ "ns"
  ))

# === STEP 7: SV Rate Plot ===
p_sv <- ggplot(sv_summary, aes(x = direction, y = mean_sv_rate, fill = direction)) +
  geom_boxplot(outlier.shape = NA, alpha = 0.7, width = 0.6) +
  geom_jitter(width = 0.15, alpha = 0.6, size = 2, aes(color = direction)) +
  facet_wrap(~type, scales = "free_y") +
  scale_fill_manual(values = c("Short" = "#d95f02", "Long" = "#1b9e77")) +
  scale_color_manual(values = c("Short" = "#d95f02", "Long" = "#1b9e77")) +
  theme_minimal(base_size = 14) +
  labs(
    title = "SV Rate by Telomere Length Group and SV Type",
    x = "Telomere Length Group",
    y = "Mean SV Rate (per Mb)",
    fill = "Telomere Group",
    color = "Telomere Group"
  ) +
  theme(legend.position = "none") +
  geom_text(data = sv_pvals, aes(x = 1.5, y = Inf, label = signif), 
            inherit.aes = FALSE, vjust = 1.5, size = 6)

# === STEP 8: CNV Rate Plot ===
p_cnv <- ggplot(cnv_summary, aes(x = direction, y = mean_cnv_rate, fill = direction)) +
  geom_boxplot(outlier.shape = NA, alpha = 0.7, width = 0.6) +
  geom_jitter(width = 0.15, alpha = 0.6, size = 2, aes(color = direction)) +
  facet_wrap(~Gain_Loss, scales = "free_y") +
  scale_fill_manual(values = c("Short" = "#d95f02", "Long" = "#1b9e77")) +
  scale_color_manual(values = c("Short" = "#d95f02", "Long" = "#1b9e77")) +
  theme_minimal(base_size = 14) +
  labs(
    title = "CNV Rate by Telomere Length Group and Gain/Loss",
    x = "Telomere Length Group",
    y = "Mean CNV Rate (per Mb)",
    fill = "Telomere Group",
    color = "Telomere Group"
  ) +
  theme(legend.position = "none") +
  geom_text(data = cnv_pvals, aes(x = 1.5, y = Inf, label = signif), 
            inherit.aes = FALSE, vjust = 1.5, size = 6)

# === STEP 9: Print plots ===
print(p_sv)
print(p_cnv)
```
