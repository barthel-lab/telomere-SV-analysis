```r
# Load the data
data <- read.table("BJ_filtered_telomer_lengths.txt", header = TRUE, stringsAsFactors = FALSE, sep = "\t")

# Rename the column
colnames(data) <- c("combined")

# Split the combined column into individual columns
data <- data %>%
  separate(combined, into = c("chr", "ref_samp", "TL_p75"), sep = "\\s+")

# Convert TL_p75 to numeric
data$TL_p75 <- as.numeric(data$TL_p75)

# Group by chr and select the maximum TL_p75
library(dplyr)
normal_data <- data %>%
  group_by(chr) %>%
  summarise(max_TL_p75 = max(TL_p75, na.rm = TRUE))

# Save the results to a file
write.table(normal_data, file = "IMR90_filtered_telomer_lengths.txt", row.names = FALSE, sep = "\t", quote = FALSE)

# Print the result
print(normal_data)
```

```r
# Load necessary libraries
library(dplyr)
library(data.table)

# Read the large dataset (replace 'your_file.csv' with your file path)
data <- fread("telomeres_final_data_filtered.csv")

# Summarize data: calculate the average TL for each chromosome arm (p and q)
tumor_data <- data %>%
  group_by(chr, arm) %>%
  summarise(mean_TL = mean(final_average_TL_p75, na.rm = TRUE)) %>%
  ungroup() %>%
  mutate(group = "tumor") %>%
  select(group, chr, mean_TL)  # Select only the important columns

# Save the summarized data to a CSV file
# fwrite(chromosome_average, "tumor_chromosome_average.csv")

# Print the result
print(tumor_data)
```

```r
colnames(tumor_data)
colnames(normal_data)
```

```r
library(dplyr)

# Rename the columns for tumor and normal datasets
tumor_data <- tumor_data %>%
  rename(tumor_tel_length = mean_TL)

normal_data <- normal_data %>%
  rename(normal_tel_length = max_TL_p75)

# Merge the datasets based on the 'chr' column
merged_data <- left_join(tumor_data, normal_data, by = "chr")

# View the merged data
print(merged_data)
```

```r
library(dplyr)

# Remove 'group.x' and 'group.y' columns
cleaned_data <- merged_data %>%
  select(chr, tumor_tel_length, normal_tel_length)

# View the cleaned data
print(cleaned_data)
```

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Ensure that 'group' is a factor and set the levels correctly
mean_tel_data <- cleaned_data %>%
  pivot_longer(cols = c("tumor_tel_length", "normal_tel_length"),
               names_to = "group", 
               values_to = "telomere_length") %>%
  mutate(group = factor(group, levels = c("tumor_tel_length", "normal_tel_length"), 
                        labels = c("Tumor", "Normal"))) %>%
  group_by(chr, group) %>%
  summarise(mean_tel_length = mean(telomere_length, na.rm = TRUE), .groups = "drop")

# Plot the data
ggplot(mean_tel_data, aes(x = chr, y = mean_tel_length, fill = group)) +
  geom_bar(stat = "identity", position = "dodge", color = "black", width = 0.7) +
  labs(x = "Chromosome", y = "Mean Telomere Length (p75)",
       title = "Comparison of Tumor and Normal Telomere Lengths by Chromosome") +
  scale_fill_manual(values = c("Tumor" = "#FF6347", "Normal" = "#4682B4")) +  # Custom colors (Red for Tumor, Blue for Normal)
  theme_minimal(base_size = 14) +  # Clean theme with larger text size
  theme(axis.text.x = element_text(angle = 45, hjust = 1),  # Angle X-axis labels for readability
        axis.title = element_text(face = "bold"),
        legend.position = "top")  # Move legend to the top
options(repr.plot.width = 10, repr.plot.height = 4, repr.plot.res = 200)
# Show the plot
print(group_plot)

```

```r
# Transform data to long format for ggplot
data_long <- cleaned_data %>%
  tidyr::pivot_longer(cols = c(tumor_tel_length, normal_tel_length),
                      names_to = "Type", values_to = "Telomere_Length")
```

```r
# Create the boxplot
p <- ggplot(data_long, aes(x = Type, y = Telomere_Length, fill = Type)) +
  geom_boxplot() +
  scale_fill_manual(values = c("normal_tel_length" = "darkgreen", 
                               "tumor_tel_length" = "red")) +
  labs(x = "Type", y = "Telomere Length", title = "Telomere Length Comparison") +
  theme_minimal()

options(repr.plot.width = 4, repr.plot.height = 6, repr.plot.res = 200)

# Add statistical test
p + stat_compare_means(method = "t.test", comparisons = list(c("normal_tel_length", "tumor_tel_length")),
                       label = "p.signif")
```
