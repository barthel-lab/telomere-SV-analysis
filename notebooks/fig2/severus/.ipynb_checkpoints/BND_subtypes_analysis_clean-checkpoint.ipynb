```r

# stackbar plot `non-chromotripsis` vs chromotripsis by grade


df <- tibble(
  Grade = c(3, 2),
  `non-chromotripsis` = c(3, 5),
  chromotripsis = c(6, 1)
)

df_long <- df %>%
  pivot_longer(
    cols = c(`non-chromotripsis`, chromotripsis),
    names_to = "Category",
    values_to = "Count"
  )

ggplot(df_long, aes(x = factor(Grade), y = Count, fill = Category)) +
  geom_col(color = "white") +
  scale_fill_manual(values = c("non-chromotripsis" = "grey70",
                               "chromotripsis" = "firebrick")) +
  labs(x = "Grade", y = "Sample count", fill = "") +
  theme_classic(base_size = 14)
```

```r


# 1. Data Setup
df_counts <- tibble(
  Grade = c("Grade 2", "Grade 3"),
  `Non-Chromothripsis` = c(5, 3),
  Chromothripsis = c(1, 6)
)

# 2. Statistical Validation (Fisher's Exact Test)
# Matrix matches the data: Grade 2 (5,1), Grade 3 (3,6)
contingency_matrix <- matrix(c(5, 1, 3, 6), nrow = 2, byrow = TRUE)
p_val <- fisher.test(contingency_matrix)$p.value

# 3. Reshape for Plotting
df_long <- df_counts %>%
  pivot_longer(cols = -Grade, names_to = "Status", values_to = "Count") %>%
  mutate(Grade = factor(Grade, levels = c("Grade 2", "Grade 3")),
         Status = factor(Status, levels = c("Non-Chromothripsis", "Chromothripsis")))

# 4. Upgraded Plot
p <- ggplot(df_long, aes(x = Grade, y = Count, fill = Status)) +
  geom_col(color = "black", width = 0.5, size = 0.3) + # Thin black outline for professional look
  scale_fill_manual(values = c("Non-Chromothripsis" = "#BDC3C7", 
                               "Chromothripsis" = "#FADADD")) + # Soft pink
  
  # Significance Bracket
  geom_signif(
    y_position = 9.5, xmin = 1, xmax = 2,
    annotation = paste0("p = ", round(p_val, 3)), 
    tip_length = 0.05, textsize = 4, fontface = "italic"
  ) +
  
  # Visual Styling
  scale_y_continuous(breaks = seq(0, 10, 2), limits = c(0, 11)) +
  scale_x_discrete(expand = c(0.4, 0.4)) + 
  labs(
    title = "Chromothripsis Prevalence Increases with Grade",
    subtitle = "IDH-mutant Astrocytoma Cohort (n=15)",
    x = NULL, 
    y = "Number of Arm Events", 
    fill = "Genomic Status"
  ) +
  theme_classic() +
  theme(
    legend.position = "bottom",
    legend.title = element_text(size = 10, face = "bold"),
    plot.title = element_text(face = "bold", size = 14, hjust = 0.5),
    plot.subtitle = element_text(size = 10, hjust = 0.5, color = "grey30"),
    axis.text = element_text(size = 12, color = "black"),
    axis.title.y = element_text(size = 12),
    aspect.ratio = 1.3 # Taller plot looks better in multi-panel figures
  )

print(p)
```

```r
# --- 1.  GRADE ANNOTATION for heatmap plot---
clinical_anno <- master_df %>%
  select(Grade) %>%
  mutate(Grade = as.character(Grade)) %>%
  mutate(Grade = ifelse(is.na(Grade) | Grade == "NA", "Unknown", Grade))

#  Grade Colors
col_list = list(
  Grade = c("2" = "red", "3" = "darkred")
)

#heatmap left side annotation
left_ha = rowAnnotation(
  Grade = clinical_anno$Grade,
  col = col_list,
  simple_anno_size = unit(4, "mm"),
  annotation_label = "Grade",
  show_annotation_name = TRUE
)

# --- 2. SUMMARY ANNOTATIONS (side barplots with heatmap) for frequency of subtypes ---
max_val <- max(mat, na.rm = TRUE)
col_fun = colorRamp2(c(0, 1, max_val), c("#ffffff", "#deebf7", "#08306b"))

top_ha = HeatmapAnnotation(
  "Type Frequency" = anno_barplot(colSums(mat), 
                                  gp = gpar(fill = "gray30", col = NA),
                                  border = FALSE),
  height = unit(1, "cm")
)

#right side annotation
right_ha = rowAnnotation(
  "Total Burden" = anno_barplot(rowSums(mat), 
                                gp = gpar(fill = "darkred", col = NA),
                                border = FALSE),
  width = unit(2, "cm")
)
```

```r
# --- 4. heatmap ---
p = Heatmap(mat, 
        name = "Events\nCount", 
        col = col_fun,
        rect_gp = gpar(col = "white", lwd = 0.5), # White grid for "clean" look
        cluster_rows = TRUE, 
        cluster_columns = FALSE,
        clustering_method_rows = "ward.D2",
        left_annotation = left_ha,   
        top_annotation = top_ha,      
        right_annotation = right_ha, 
        column_title = "Complex BND subtypes",
        column_title_gp = gpar(fontsize = 14, fontface = "bold"),
        column_names_rot = 90,
        row_names_gp = gpar(fontsize = 7),
        column_names_gp = gpar(fontsize = 10, fontface = "bold"))

# Draw the final result
draw(p, merge_legends = TRUE, heatmap_legend_side = "right", annotation_legend_side = "right")
```

```r
# 3. Create Plot A: Top Bar Chart (ecDNA Fragments)
p_top <- ggplot(master_df, aes(x = reorder(Sample, -ecDNA), y = ecDNA)) +
  geom_bar(stat = "identity", fill = "firebrick", width = 0.7) +
  theme_minimal() +
  theme(
    axis.title.x = element_blank(),
    axis.text.x = element_blank(),
    axis.ticks.x = element_blank(),
    panel.grid.major.x = element_blank()
  ) +
  labs(y = "ecDNA Frags", title = "ecDNA Burden & SV Association Profile")

# 4. Create Plot B: Bottom Bubble Matrix (SV Types)
p_bottom <- ggplot(df_bubbles, aes(x = reorder(Sample, -ecDNA), y = SV_Type)) +
  # Size shows event count, Color shows ecDNA fragment intensity
  geom_point(aes(size = Event_Count, color = ecDNA), alpha = 0.7) +
  scale_size_continuous(range = c(0, 15), name = "SV Events") + 
  scale_color_gradient(low = "yellow", high = "yellow3", name = "ecDNA Count") +
  theme_bw() +
  theme(
    axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1),
    panel.grid.minor = element_blank()
  ) +
  labs(x = "Samples (Sorted by ecDNA Fragments)", y = "SV Type")

# 5. Combine the plots using patchwork
# 
combined_plot <- p_top / p_bottom + plot_layout(heights = c(1, 3))

# Display the plot
combined_plot
```

```r
# 2. Updated Plot B: Bottom Bubble Matrix (SV Types)
p_bottom <- ggplot(df_bubbles, aes(x = reorder(Sample, -ecDNA), y = SV_Type)) +
  # Use subset() inside geom_point to remove zero-count bubbles
  geom_point(data = subset(df_bubbles, Event_Count > 0), 
             aes(size = Event_Count, fill = ecDNA), 
             color = "black", 
             shape = 21, 
             alpha = 0.8) +
  scale_size_continuous(range = c(2, 14), name = "SV Events") + 
  # High-impact Golden Gradient
  scale_fill_gradient(low = "#FFF9E6", high = "#D4AF37", name = "ecDNA Count") + 
  theme_bw() +
  theme(
    axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1),
    panel.grid.minor = element_blank(),
    # Remove grid lines if you want a cleaner "matrix" look
    panel.grid.major = element_line(color = "grey95") 
  ) +
  labs(x = "Samples (Sorted by ecDNA Burden)", y = "Structural Variant Type")

# 3. Combine with Patchwork
combined_plot <- (p_top / p_bottom) + 
  plot_layout(heights = c(1, 2.5)) &
  theme(legend.position = "right")

# Display
combined_plot
```

```r
library(psych)
library(pheatmap)
library(RColorBrewer)
library(dplyr)

# 1. Prepare Numeric Matrix
assoc_data <- master_df %>%
  mutate(Grade = as.numeric(as.character(Grade))) %>%
  select(all_of(names(df_columns))) %>%
  as.matrix()

# Apply clean names from your mapping
colnames(assoc_data) <- df_columns[colnames(assoc_data)]

# 2. Compute Spearman Association
# Note: corr.test calculates p-values and applies adjustment
cor_results <- corr.test(assoc_data, method = "spearman", adjust = "fdr")

# --- SYMMETRY FIX START ---
# Extract the correlation (r) and p-values (p)
r_matrix <- cor_results$r
p_matrix <- cor_results$p

# Force symmetry: ensure the upper triangle matches the lower triangle
# This fixes the "different stars on different sides" issue
p_matrix[upper.tri(p_matrix)] <- t(p_matrix)[upper.tri(p_matrix)]
r_matrix[upper.tri(r_matrix)] <- t(r_matrix)[upper.tri(r_matrix)]
# --- SYMMETRY FIX END ---

# 3. Significance Matrix (Stars)
# Using the corrected p_matrix
sig_matrix <- matrix(ifelse(p_matrix < 0.001, "***", 
                     ifelse(p_matrix < 0.01, "**", 
                     ifelse(p_matrix < 0.05, "*", ""))), 
                     nrow = nrow(r_matrix))

# 4. Heatmap Generation
# We use 'r_matrix' and 'sig_matrix' which are now perfectly mirrored
pheatmap(r_matrix, 
         display_numbers = sig_matrix,
         cluster_rows = TRUE, 
         cluster_cols = TRUE,
         clustering_method = "ward.D2", 
         color = colorRampPalette(rev(brewer.pal(n = 11, name = "RdBu")))(100),
         main = "Spearman Association: SV Rates, Proportions, and ecDNA\n(FDR Adjusted)",
         fontsize_number = 10,
         fontsize_row = 9,
         fontsize_col = 9,
         number_color = "black",
         border_color = "white",
         cellwidth = 30, 
         cellheight = 30,
         angle_col = 45)
```

```r
library(psych)
library(pheatmap)
library(RColorBrewer)
library(dplyr)
library(grid)

# 1. Prepare Numeric Matrix
assoc_data <- master_df %>%
  mutate(Grade = as.numeric(as.character(Grade))) %>%
  select(all_of(names(df_columns))) %>%
  as.matrix()

colnames(assoc_data) <- df_columns[colnames(assoc_data)]

# 2. Compute Spearman Association & Fix Symmetry
cor_results <- corr.test(assoc_data, method = "spearman", adjust = "fdr")

r_matrix <- cor_results$r
p_matrix <- cor_results$p

# Force Mirror Symmetry (Essential for Spearman)
p_matrix[upper.tri(p_matrix)] <- t(p_matrix)[upper.tri(p_matrix)]
r_matrix[upper.tri(r_matrix)] <- t(r_matrix)[upper.tri(r_matrix)]

# 3. Significance Stars (Cleaned up for publication)
sig_matrix <- matrix(ifelse(p_matrix < 0.001, "***", 
                     ifelse(p_matrix < 0.01, "**", 
                     ifelse(p_matrix < 0.05, "*", ""))), 
                     nrow = nrow(r_matrix))

# 4. Nature-Style Heatmap
# Note: We use a specific color break to make '0' correlation pure white
breaksList = seq(-1, 1, by = 0.02)
my_color = colorRampPalette(rev(brewer.pal(n = 7, name = "RdBu")))(length(breaksList))

#pdf("Nature_Genetics_ecDNA_Heatmap.pdf", width = 7, height = 6) # Export as PDF for Vector Graphics

pheatmap(r_matrix, 
         display_numbers = sig_matrix,
         cluster_rows = TRUE, 
         cluster_cols = TRUE,
         clustering_method = "ward.D2", 
         color = my_color,
         breaks = breaksList,
         main = "", # Titles are usually put in the Figure Legend, not on the plot
         fontsize_number = 8,  # Smaller, cleaner stars
         fontsize_row = 10,    # Arial-like sizing
         fontsize_col = 10,
         number_color = "black",
         border_color = "white", # Thin white borders define the "Nature" look
         cellwidth = 22, 
         cellheight = 22,
         treeheight_row = 30,  # Clearer dendrograms
         treeheight_col = 30,
         angle_col = 45,
         legend_labels = c("-1", "0", "1"),
         legend_title = expression(paste("Spearman ", rho)))

#dev.off()
```
