```r
library(tidyverse)      # Includes: dplyr, ggplot2, tidyr, readr, stringr, purrr, forcats
library(tibble)
library(scales)         # For axis formatting
library(GenomicRanges)
library(IRanges)
library(lme4)           # Linear mixed-effects models
library(lmerTest)       # P-values for lmer
library(survival)       # Survival analysis
library(survminer)      # Visualization of survival curves
library(gridExtra)      # Side-by-side plots
library(grid)           # Low-level plot adjustments
library(ggforce)        # For geom_sina and advanced shapes
library(ggpubr)         # For stat_compare_means and easy annotations
library(ggsci)          # Nature/Science journal color palettes
library(ComplexHeatmap) # Advanced heatmaps
library(circlize)       # Required for circular layouts/ComplexHeatmap colors
library(forcats)
```

```r
library(readr)
violin_df <- read_tsv("/home/mjehangir/telomere-sv-analysis/data/TEL_merged_HARs_with_chr-arms.txt")

# View the first few rows
head(violin_df)
```

```r

# Ensure proper labels/order
violin_df <- violin_df %>%
  mutate(
    ecDNA_flag = case_when(
      ecDNA_flag %in% c(1, "1", "ecDNA") ~ "ecDNA",
      TRUE                                ~ "No ecDNA"
    ),
    ecDNA_flag = factor(ecDNA_flag, levels = c("No ecDNA", "ecDNA"))
  )

# Palettes
fill_pal  <- c("No ecDNA" = "grey", "ecDNA" = "gold")
point_pal <- c("No ecDNA" = "grey", "ecDNA" = "#b8860b")  # darker outline for contrast

ymax <- max(violin_df$final_average_TL_p75, na.rm = TRUE)

pB <- ggplot(violin_df, aes(x = ecDNA_flag, y = final_average_TL_p75, fill = ecDNA_flag)) +
  # Violin & box
  geom_violin(trim = FALSE, alpha = 0.5, color = NA, width = 0.9) +
  # Mean marker
  #stat_summary(fun = mean, geom = "point", shape = 23, size = 2.4,
   #            fill = "white", color = "black") +
  # SINAPLOT points (density-based jitter)
  geom_sina(aes(color = ecDNA_flag), size = 1.4, alpha = 0.6, maxwidth = 0.7) +
    geom_boxplot(width = 0.40, outlier.shape = NA, alpha = 0.85, color = "black") +
  # Scales
  scale_fill_manual(values = fill_pal, guide = "none") +
  scale_color_manual(values = point_pal, guide = "none") +
  # Labels/theme
  labs(x = NULL, y = "Telomere Length (p75)") +
  theme_classic(base_size = 12) +
  theme(
    axis.text.x  = element_text(size = 11),
    axis.title.y = element_text(size = 12)
  ) +
  # Significance (Wilcoxon)
  stat_compare_means(
    aes(label = ..p.format..),
    method = "wilcox.test",
    comparisons = list(c("No ecDNA", "ecDNA")),
    label.y = 1.05 * ymax
  )
```

```r
options(repr.plot.width = 3.5, repr.plot.height = 4, repr.plot.res = 200)

print(pB)
```

```r

# ---- 1) Aggregation (Remove Pseudoreplication) ----
violin_df_clean <- violin_df %>%
  filter(!is.na(final_average_TL_p75)) %>%
  group_by(Sample_prefix, chr_arm) %>%
  summarise(
    # Take the first TL value since it's identical for all segments on this arm
    final_average_TL_p75 = dplyr::first(final_average_TL_p75),
    # If any segment on this arm is ecDNA, the whole arm counts as ecDNA
    ecDNA_status = if_else(any(ecDNA_flag %in% c(1, "1", "ecDNA")), "ecDNA", "No ecDNA"),
    .groups = "drop"
  ) %>%
  mutate(ecDNA_status = factor(ecDNA_status, levels = c("No ecDNA", "ecDNA")))

# ---- 2) Prepare Labels with n counts ----
counts <- table(violin_df_clean$ecDNA_status)
x_labels <- c(
  paste0("No ecDNA\n(n=", counts["No ecDNA"], ")"),
  paste0("ecDNA\n(n=", counts["ecDNA"], ")")
)

# ---- 3) Plotting ----
ymax <- max(violin_df_clean$final_average_TL_p75, na.rm = TRUE)

ggplot(violin_df_clean, aes(x = ecDNA_status, y = final_average_TL_p75, fill = ecDNA_status)) +
  # Geoms
  geom_violin(trim = FALSE, alpha = 0.4, color = NA, width = 0.8) +
  geom_boxplot(width = 0.2, outlier.shape = NA, alpha = 0.8, fill = "white") +
  geom_sina(aes(color = ecDNA_status), size = 1.2, alpha = 0.5) +
  # Scales & Themes
  scale_fill_manual(values = c("No ecDNA" = "grey70", "ecDNA" = "gold")) +
  scale_color_manual(values = c("No ecDNA" = "grey40", "ecDNA" = "#b8860b")) +
  scale_x_discrete(labels = x_labels) +
  labs(x = NULL, y = "Telomere Length (p75)", title = "Arm-Level Telomere Analysis") +
  theme_classic(base_size = 12) +
  theme(legend.position = "none", plot.title = element_text(hjust = 0.5, face = "bold")) +
  # Statistics
  stat_compare_means(
    method = "wilcox.test",
    comparisons = list(c("No ecDNA", "ecDNA")),
    label.y = 1.1 * ymax
  )
```

```r
library(lmerTest)

# 1. Prepare Data (One row per segment, but with grouping)
mixed_df <- violin_df %>%
  filter(!is.na(final_average_TL_p75)) %>%
  mutate(
    ecDNA_status = factor(ifelse(ecDNA_flag %in% c(1, "1", "ecDNA"), "ecDNA", "No ecDNA"),
                          levels = c("No ecDNA", "ecDNA"))
  )

# 2. Run a Simpler Mixed Model
# We use (1 | Sample_prefix) to account for patient-level correlation.
# This avoids the "VtV" error by not over-nesting.
model_simple <- lmer(final_average_TL_p75 ~ ecDNA_status + (1 | Sample_prefix), data = mixed_df)

# Extract P-value
stats_summary <- summary(model_simple)
p_val <- stats_summary$coefficients["ecDNA_statusecDNA", "Pr(>|t|)"]
p_label <- paste0("Mixed Model (by Sample) p = ", signif(p_val, 3))

# 3. Plotting
ymax <- max(mixed_df$final_average_TL_p75, na.rm = TRUE)

ggplot(mixed_df, aes(x = ecDNA_status, y = final_average_TL_p75, fill = ecDNA_status)) +
  geom_violin(trim = FALSE, alpha = 0.4, color = NA) +
  geom_boxplot(width = 0.2, outlier.shape = NA, fill = "white") +
  geom_sina(aes(color = ecDNA_status), size = 1.2, alpha = 0.3) +
  scale_fill_manual(values = c("No ecDNA" = "grey70", "ecDNA" = "gold")) +
  scale_color_manual(values = c("No ecDNA" = "grey40", "ecDNA" = "#b8860b")) +
  labs(
    x = NULL, 
    y = "Telomere Length (p75)", 
    subtitle = p_label,
    title = "Telomere Length Corrected for Sample Clustering"
  ) +
  theme_classic(base_size = 12) +
  theme(legend.position = "none", plot.title = element_text(hjust = 0.5, face = "bold"))
```
