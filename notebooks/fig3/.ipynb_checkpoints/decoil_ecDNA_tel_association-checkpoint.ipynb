```r
library(tidyverse)      # Includes: dplyr, ggplot2, tidyr, readr, stringr, purrr, forcats
library(tibble)
library(scales)         # For axis formatting
library(GenomicRanges)
library(IRanges)
library(lme4)           # Linear mixed-effects models
library(lmerTest)       # P-values for lmer
library(survival)       # Survival analysis
library(survminer)      # Visualization of survival curves
library(gridExtra)      # Side-by-side plots
library(grid)           # Low-level plot adjustments
library(ggforce)        # For geom_sina and advanced shapes
library(ggpubr)         # For stat_compare_means and easy annotations
library(ggsci)          # Nature/Science journal color palettes
library(ComplexHeatmap) # Advanced heatmaps
library(circlize)       # Required for circular layouts/ComplexHeatmap colors
library(forcats)
```

```r
#read the ecDNA dataset
df_ec <- read.csv("ecDNA_arm_level.csv", header = TRUE, sep = ',')

df_ec = df_ec %>% select(sample, chr_arm, total_prop, n_fragments, avg_TL_kb, avg_norm_TL)

head(df_ec)
```

```r
nrow(df_ec)
```

```r
# ---- Compare: Absolute TL vs Total ecDNA Burden association ----

options(repr.plot.width = 5, repr.plot.height = 5, repr.plot.res = 200)

p_abs <- ggplot(df_ec, aes(x = total_prop, y = avg_TL_kb)) +
  
# Plot points - size them by fragment count to show complexity
  geom_point(aes(size = n_fragments), alpha = 0.6, color = "#E69F00") + 
  
  # Add the regression line
  geom_smooth(method = "lm", color = "black", fill = "grey80", se = TRUE) +
  
  # Pearson correlation
  stat_cor(method = "pearson", size = 5, label.x.npc = "left") +
  
  labs(
    title = "ecDNA Burden vs. Telomere Length",
    subtitle = "Aggregated by chromosome arm",
    x = "Total ecDNA Proportion (per arm)",
    y = "Telomere Length (kb)",
    size = "Fragments"
  ) +
  theme_classic(base_size = 14) +
  theme(
    axis.text = element_text(color = "black"),
    plot.title = element_text(face = "bold")
  )

# View the plot
p_abs
```

```r
# ---- : Normalized TL vs Total ecDNA Burden association ----

options(repr.plot.width = 5, repr.plot.height = 5, repr.plot.res = 200)

p_norm <- ggplot(df_ec, aes(x = total_prop, y = avg_norm_TL)) +
  
  # Plot points - size them by fragment count to show complexity
  geom_point(aes(size = n_fragments), alpha = 0.6, color = "#E69F00") + 

  geom_smooth(method = "lm", color = "black", fill = "grey80", se = TRUE) +
  
  # Pearson correlation 
  stat_cor(method = "pearson", size = 5, label.x.npc = "left") +
  
  labs(
    title = "ecDNA Burden vs. Normalized telomere length",
    subtitle = "Aggregated by chromosome arm",
    x = "Total ecDNA Proportion (per arm)",
    y = "Telomere Length (kb)",
    size = "Fragments"
  ) +
  theme_classic(base_size = 14) +
  theme(
    axis.text = element_text(color = "black"),
    plot.title = element_text(face = "bold")
  )

# View the plot
p_norm
```
