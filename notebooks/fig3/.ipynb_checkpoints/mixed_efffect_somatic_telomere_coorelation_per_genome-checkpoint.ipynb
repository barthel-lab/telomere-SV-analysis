```r
library(dplyr)
library(lme4)
library(clubSandwich)
library(ggplot2)

extract_FE_estimates <- function(df, ycol) {

  df %>%
    mutate(
      y = log1p(as.numeric(.data[[ycol]])),
      TL_group = factor(TL_group, levels = c("Short TL", "Other")),
      Sample_ID = factor(Sample_ID),
      type = factor(type)
    ) %>%
    filter(is.finite(y)) %>%
    group_by(type) %>%
    filter(n() >= 2) %>%
    group_modify(~ {
      tryCatch({
        fit <- lmer(
         # y ~ TL_group * final_average_TL_p75 + (1 | Sample_ID),
         y ~ TL_group + (1 | Sample_ID),

          data = .x
        )

        ct <- coef_test(
          fit,
          vcov = "CR2",
          cluster = .x$Sample_ID
        )

        tibble(
          estimate = ct["TL_groupOther", "beta"],
          se       = ct["TL_groupOther", "SE"],
          p        = ct["TL_groupOther", "p_Satt"]
        )
      }, error = function(e) {
        tibble(estimate = NA_real_, se = NA_real_, p = NA_real_)
      })
    }) %>%
    ungroup() %>%
    mutate(
      conf.low  = estimate - 1.96 * se,
      conf.high = estimate + 1.96 * se,
      signif = case_when(
        p < 0.001 ~ "***",
        p < 0.01  ~ "**",
        p < 0.05  ~ "*",
        TRUE ~ "ns"
      )
    )
}
```

```r
make_forest_plot <- function(df, title_text) {

  ggplot(df, aes(x = estimate, y = type)) +
    geom_vline(xintercept = 0, linetype = "dashed", color = "grey40") +
    geom_errorbarh(
      aes(xmin = conf.low, xmax = conf.high),
      height = 0.25,
      linewidth = 0.8,
      color = "black"
    ) +
    geom_point(
      size = 3,
      shape = 21,
      fill = "steelblue"
    ) +
    geom_text(
      aes(label = signif),
      hjust = -0.6,
      size = 4,
      fontface = "bold"
    ) +
    labs(
      title = title_text,
      subtitle = "Effect of Other vs Short TL (log1p scale)\nMixed-effects model with CR2 robust SEs",
      x = "β estimate (95% CI)",
      y = NULL
    ) +
    theme_classic(base_size = 14) +
    theme(
      plot.title = element_text(face = "bold", hjust = 0.5),
      plot.subtitle = element_text(hjust = 0.5),
      axis.text.y = element_text(face = "bold")
    )
}
```

```r
forest_sv_rate <- extract_FE_estimates(
  merged_data,
  "sv_rate"
)

p_forest_rate <- make_forest_plot(
  forest_sv_rate,
  "SV rate by TL group"
)
forest_sv_prop <- extract_FE_estimates(
  merged_data,
  "sv_proportion"
)

p_forest_prop <- make_forest_plot(
  forest_sv_prop,
  "SV proportion by TL group"
)
```

```r
# ROBUST QUANTILE REGRESSION WITH BOOTSTRAP CONFIDENCE INTERVALS
# Uses bootstrap to get stable confidence intervals

library(tidyverse)
library(quantreg)
library(ggplot2)

# ============================================================================
# 0. DATA DIAGNOSTICS
# ============================================================================

cat("\n========== DATA DIAGNOSTICS ==========\n")

cat("\nTotal observations:", nrow(merged_data))
cat("\nUnique patients:", n_distinct(merged_data$Sample_ID))
cat("\nTL_group distribution:\n")
print(table(merged_data$TL_group))

sv_breakdown <- merged_data %>%
  group_by(type, TL_group) %>%
  summarise(
    n = n(),
    n_patients = n_distinct(Sample_ID),
    median_rate = median(sv_rate, na.rm = TRUE),
    mean_rate = mean(sv_rate, na.rm = TRUE),
    median_prop = median(sv_proportion, na.rm = TRUE),
    mean_prop = mean(sv_proportion, na.rm = TRUE),
    .groups = "drop"
  )

print("\nSV Type breakdown (with medians):")
print(sv_breakdown)

# ============================================================================
# 1. QUANTILE REGRESSION WITH BOOTSTRAP CI
# ============================================================================

sv_types <- unique(merged_data$type)
results_all <- tibble()

for (sv in sv_types) {
  
  cat("\n=== SV Type:", sv, "===\n")
  
  data_sv <- merged_data %>% filter(type == sv)
  
  # ========== QUANTILE REGRESSION 1: SV Rate (MEDIAN) ==========
  tryCatch({
    # Fit quantile regression with bootstrap CIs
    qr_rate <- rq(
      sv_rate ~ TL_group,
      data = data_sv,
      tau = 0.5
    )
    
    # Get bootstrap confidence intervals
    qr_rate_ci <- summary(qr_rate, se = "boot", bsmethod = "xy")
    
    coef_rate <- coef(qr_rate)
    coef_names <- names(coef_rate)
    
    # Extract only TL_group term
    tl_term <- coef_names[grep("TL_group", coef_names)][1]
    
    if (!is.na(tl_term)) {
      estimate <- coef_rate[tl_term]
      se_val <- qr_rate_ci$coefficients[tl_term, "Std. Error"]
      pval <- qr_rate_ci$coefficients[tl_term, "Pr(>|t|)"]
      
      results_rate <- tibble(
        sv_type = sv,
        metric = "SV Rate",
        estimate = estimate,
        std.error = se_val,
        p.value = pval,
        ci_lower = estimate - 1.96 * se_val,
        ci_upper = estimate + 1.96 * se_val,
        significant = pval < 0.05
      )
      
      results_all <- bind_rows(results_all, results_rate)
      cat("  ✓ SV Rate: estimate =", round(estimate, 6), "p =", round(pval, 4), "\n")
    }
    
  }, error = function(e) {
    cat("  ✗ SV Rate model failed\n")
  })
  
  # ========== QUANTILE REGRESSION 2: SV Proportion (MEDIAN) ==========
  tryCatch({
    qr_prop <- rq(
      sv_proportion ~ TL_group,
      data = data_sv,
      tau = 0.5
    )
    
    # Get bootstrap confidence intervals
    qr_prop_ci <- summary(qr_prop, se = "boot", bsmethod = "xy")
    
    coef_prop <- coef(qr_prop)
    coef_names <- names(coef_prop)
    
    # Extract only TL_group term
    tl_term <- coef_names[grep("TL_group", coef_names)][1]
    
    if (!is.na(tl_term)) {
      estimate <- coef_prop[tl_term]
      se_val <- qr_prop_ci$coefficients[tl_term, "Std. Error"]
      pval <- qr_prop_ci$coefficients[tl_term, "Pr(>|t|)"]
      
      results_prop <- tibble(
        sv_type = sv,
        metric = "SV Proportion",
        estimate = estimate,
        std.error = se_val,
        p.value = pval,
        ci_lower = estimate - 1.96 * se_val,
        ci_upper = estimate + 1.96 * se_val,
        significant = pval < 0.05
      )
      
      results_all <- bind_rows(results_all, results_prop)
      cat("  ✓ SV Proportion: estimate =", round(estimate, 8), "p =", round(pval, 4), "\n")
    }
    
  }, error = function(e) {
    cat("  ✗ SV Proportion model failed\n")
  })
}

# ============================================================================
# 2. PREPARE DATA FOR FOREST PLOT
# ============================================================================

plot_data <- results_all %>%
  mutate(
    line_label = paste0(sv_type, " - ", metric),
    line_label = fct_reorder(line_label, estimate)
  )

print("\n========== QUANTILE REGRESSION RESULTS (Bootstrap CI) ==========")
print(plot_data %>% 
  select(sv_type, metric, estimate, p.value, significant))

# ============================================================================
# 3. FOREST PLOT - QUANTILE REGRESSION (Robust Method)
# ============================================================================
options(repr.plot.width = 7, repr.plot.height = 4)  # reset if needed

forest_plot <- plot_data %>%
  ggplot(aes(y = line_label, x = estimate, color = significant, size = significant)) +
  
  # Vertical line at 0 (no effect)
  geom_vline(xintercept = 0, linetype = "dashed", color = "gray50", linewidth = 1) +
  
  # Error bars (95% confidence intervals from bootstrap)
  geom_errorbarh(aes(xmin = ci_lower, xmax = ci_upper), 
                 height = 0.4, alpha = 0.8, linewidth = 1) +
  
  # Points for effect sizes
  geom_point(alpha = 0.9) +
  
  # Colors: Red = significant, Gray = not significant
  scale_color_manual(values = c("FALSE" = "gray60", "TRUE" = "#e74c3c"), guide = "none") +
  scale_size_manual(values = c("FALSE" = 3, "TRUE" = 5), guide = "none") +
  
  # Labels
  labs(
    title = "Quantile Regression (Median): Effect of SHORT Telomeres on SV Burden",
    subtitle = "Robust to outliers | Bootstrap 95% CI | Red = significant (p < 0.05)\nPositive = higher median in SHORT | Negative = lower median in SHORT",
    x = "Median Difference (SHORT - OTHER)",
    y = NULL
  ) +
  
  theme_minimal(base_size = 12) +
  theme(
    plot.title = element_text(size = 14, face = "bold", hjust = 0.5),
    plot.subtitle = element_text(size = 10, hjust = 0.5),
    axis.title.x = element_text(face = "bold", size = 12),
    axis.text.y = element_text(size = 11, face = "bold"),
    panel.grid.major.x = element_line(color = "gray90", linewidth = 0.3)
  )

print(forest_plot)

# ============================================================================
# 4. SUMMARY TABLE
# ============================================================================

summary_table <- plot_data %>%
  mutate(
    direction = if_else(estimate > 0, "↑ Higher in SHORT", "↓ Lower in SHORT"),
    p.value = format(p.value, scientific = TRUE, digits = 3),
    estimate = format(round(estimate, 8), scientific = TRUE),
    ci_lower = format(round(ci_lower, 8), scientific = TRUE),
    ci_upper = format(round(ci_upper, 8), scientific = TRUE)
  ) %>%
  select(sv_type, metric, direction, estimate, ci_lower, ci_upper, p.value, significant) %>%
  rename(
    "SV Type" = sv_type,
    "Metric" = metric,
    "Direction" = direction,
    "Median Effect" = estimate,
    "CI Lower" = ci_lower,
    "CI Upper" = ci_upper,
    "P-Value" = p.value,
    "Significant" = significant
  )

print("\n========== SUMMARY TABLE: QUANTILE REGRESSION ==========")
print(summary_table)

# ============================================================================
# 5. MEAN vs MEDIAN COMPARISON
# ============================================================================

comparison_data <- merged_data %>%
  group_by(type, TL_group) %>%
  summarise(
    mean_rate = mean(sv_rate, na.rm = TRUE),
    median_rate = median(sv_rate, na.rm = TRUE),
    mean_prop = mean(sv_proportion, na.rm = TRUE),
    median_prop = median(sv_proportion, na.rm = TRUE),
    n = n(),
    .groups = "drop"
  )

cat("\n========== MEAN vs MEDIAN Comparison ==========")
print("(Medians are more robust to outliers)")
print(comparison_data)

```

```r
# ============================================================================
# FIGURE 3: Mixed-Effects Models (Short TL vs Other)
# ============================================================================
library(lme4)
library(lmerTest)       # Better p-values
library(broom.mixed)    # tidy()
library(dplyr)
library(ggplot2)

# -----------------------------
# SV types
# -----------------------------
sv_types <- unique(merged_data$type)

# -----------------------------
# Function to fit mixed model safely
# -----------------------------
get_mixed_model <- function(var) {
  fig_list <- lapply(sv_types, function(sv) {
    
    dat <- subset(merged_data, type == sv)
    
    n_short <- sum(dat$TL_group == "Short TL")
    n_other <- sum(dat$TL_group == "Other")
    
    # Skip if insufficient Short TL rows
    if(n_short < 5) {
      message("Skipping ", sv, ": insufficient Short TL rows (", n_short, ")")
      return(NULL)
    }
    
    dat$TL_group <- factor(dat$TL_group, levels = c("Other", "Short TL"))
    
    # -----------------------------
    # Fit mixed model safely (ignore warnings)
    # -----------------------------
    mod <- tryCatch({
      lmer(as.formula(paste0(var, " ~ TL_group + (1 | Sample_ID)")), data = dat)
    }, error = function(e) {
      message("Error fitting ", sv, ": ", e$message)
      return(NULL)
    })
    
    if(is.null(mod)) return(NULL)
    
    # -----------------------------
    # Extract fixed effect for Short TL
    # -----------------------------
    tidy(mod, effects = "fixed", conf.int = TRUE) %>%
      filter(term == "TL_groupShort TL") %>%
      mutate(
        Variable = paste0(sv, " [", n_short, "/", n_other, "]"),
        Outcome = var
      )
  })
  
  bind_rows(fig_list)
}

# -----------------------------
# Get results for SV rate and proportion
# -----------------------------
rate_res <- get_mixed_model("sv_rate")
prop_res <- get_mixed_model("sv_proportion")

# -----------------------------
# Function to plot forest plot
# -----------------------------
plot_forest <- function(df, xlab, title_text) {
  df <- df %>%
    arrange(desc(estimate)) %>%
    mutate(Variable = factor(Variable, levels = unique(Variable)))
  
  ggplot(df, aes(x = estimate, y = Variable)) +
    geom_point(size = 3, color = "blue") +
    geom_errorbarh(aes(xmin = conf.low, xmax = conf.high), height = 0.2, color = "blue") +
    geom_vline(xintercept = 0, linetype = "dashed") +
    labs(
      x = xlab,
      y = NULL,
      title = title_text
    ) +
    theme_bw(base_size = 13)
}

# -----------------------------
# Plot SV rate
# -----------------------------
forest_rate <- plot_forest(rate_res,
                           xlab = "Effect size (Short TL vs Other) - SV rate",
                           title_text = "Figure 3a: SV Rate vs Telomere Length Group")
print(forest_rate)

# -----------------------------
# Plot SV proportion
# -----------------------------
forest_prop <- plot_forest(prop_res,
                           xlab = "Effect size (Short TL vs Other) - SV proportion",
                           title_text = "Figure 3b: SV Proportion vs Telomere Length Group")
print(forest_prop)
```

```r
library(dplyr)
library(broom)
library(ggplot2)

# -----------------------------
# 1️⃣ Average per patient per SV type
# -----------------------------
patient_avg <- merged_data %>%
  group_by(Sample_ID, TL_group, type) %>%
  summarise(
    mean_sv_rate = mean(sv_rate, na.rm = TRUE),        # use median instead of mean
    mean_sv_proportion = mean(sv_proportion, na.rm = TRUE), # same here
    .groups = "drop"
  )


sv_types <- unique(patient_avg$type)

# -----------------------------
# 2️⃣ Function to fit simple linear model per SV type
# -----------------------------
get_lm_model <- function(var) {
  fig_list <- lapply(sv_types, function(sv) {
    
    dat <- subset(patient_avg, type == sv)
    
    n_short <- sum(dat$TL_group == "Short TL")
    n_other <- sum(dat$TL_group == "Other")
    
    # Skip if too few Short TL patients
    if(n_short < 2) {
      message("Skipping ", sv, ": insufficient Short TL patients (", n_short, ")")
      return(NULL)
    }
    
    dat$TL_group <- factor(dat$TL_group, levels = c("Other", "Short TL"))
    
    mod <- tryCatch({
      lm(as.formula(paste0(var, " ~ TL_group")), data = dat)
    }, error = function(e) {
      message("Error fitting ", sv, ": ", e$message)
      return(NULL)
    })
    
    if(is.null(mod)) return(NULL)
    
    tidy(mod, conf.int = TRUE) %>%
      filter(term == "TL_groupShort TL") %>%
      mutate(
        Variable = paste0(sv, " [", n_short, "/", n_other, "]"),
        Outcome = var
      )
  })
  
  bind_rows(fig_list)
}

# -----------------------------
# 3️⃣ Get results for SV rate and proportion
# -----------------------------
rate_res <- get_lm_model("mean_sv_rate")
prop_res <- get_lm_model("mean_sv_proportion")

# -----------------------------
# 4️⃣ Function to plot forest plot
# -----------------------------
plot_forest <- function(df, xlab, title_text) {
  df <- df %>%
    arrange(desc(estimate)) %>%
    mutate(Variable = factor(Variable, levels = unique(Variable)))
  
  ggplot(df, aes(x = estimate, y = Variable)) +
    geom_point(size = 3, color = "blue") +
    geom_errorbarh(aes(xmin = conf.low, xmax = conf.high), height = 0.2, color = "blue") +
    geom_vline(xintercept = 0, linetype = "dashed") +
    labs(
      x = xlab,
      y = NULL,
      title = title_text
    ) +
    theme_bw(base_size = 13)
}

# -----------------------------
# 5️⃣ Plot SV rate
# -----------------------------
forest_rate <- plot_forest(rate_res,
                           xlab = "Effect size (Short TL vs Other) - mean SV rate",
                           title_text = "Figure 3a: Patient-Averaged SV Rate vs Telomere Length Group")
print(forest_rate)

# -----------------------------
# 6️⃣ Plot SV proportion
# -----------------------------
forest_prop <- plot_forest(prop_res,
                           xlab = "Effect size (Short TL vs Other) - mean SV proportion",
                           title_text = "Figure 3b: Patient-Averaged SV Proportion vs Telomere Length Group")
print(forest_prop)
```
